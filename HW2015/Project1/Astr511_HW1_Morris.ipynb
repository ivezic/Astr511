### Import the data

```python
%matplotlib inline
import numpy as np
import matplotlib
matplotlib.rcParams['font.size'] = 18
from matplotlib import pyplot as plt

datafile = 'Astr511HW1data.dat'
# Read in names of each column and the raw data
columnlabels = open(datafile, 'r').read().splitlines()[0][1:].split()
bigtable = np.genfromtxt(datafile)

# Make dictionary of the data where keys = column labels
data = {}
for label, column in zip(columnlabels, bigtable.T):
    data[label] = column
```

### Equation for metallicity:

$$ [\text{Fe/H}] = A + Bx + Cy + Dxy + Ex^2 + Fy^2 + Gx^2y + Hxy^2 + Ix^3 + Jy^3 \tag{1}$$

where $x = (u-g)$ and $y = (g-r)$.

### Equations from [Ivezic+ 2008](http://adsabs.harvard.edu/abs/2008ApJ...684..287I)

$$ \Delta M_r([\text{Fe/H}]) = 4.50 - 1.11[\text{Fe/H}] - 0.18[\text{Fe/H}]^2 \tag{A2}$$

$$ M_r(g-i, [\text{Fe/H}]) = M_r^0(g-i) + \Delta M_r([\text{Fe/H}]) \tag{A3}$$

$$ M_r^0(g-i) = -5.06 + 14.32x - 12.97x^2 + 6.127x^3 - 1.267x^4 + 0.0967x^5 \tag{A7}$$

where $x = (g-i)$

### Compute Distance

```python
# Compute approximate metallicity with ugr photometry as in Bond et al. 2010 (Eqn 1)
x = data['u'] - data['g']
y = data['g'] - data['r']
A, B, C, D, E, F, G, H, I, J = [-13.13, 14.09, 28.04, -5.51, -5.90, -58.68, 9.14, 
                                -20.61, 0.0, 58.20]
bondisapplicable = y < 0.6
data['metallicity'] = np.zeros_like(x)

data['metallicity'][bondisapplicable] = (A + B*x + C*y + D*x*y + E*x**2 + F*y**2 + G*y*x**2 + 
                                 H*x*y**2 + I*x**3 + J*y**3)[bondisapplicable]
data['metallicity'][np.invert(bondisapplicable)] = -0.6

# Solve Eqn A2 using the metallicity found with Eqn 1
data['DeltaM_r'] = 4.50 - 1.11*data['metallicity'] - 0.18*data['metallicity']**2

# Solve Eqn A7
x = data['g'] - data['i']
data['M_r^0'] = -5.06 + 14.32*x - 12.97*x**2 + 6.127*x**3 - 1.267*x**4 + 0.0967*x**5

# Solve Eqn A3
data['M_r'] = data['M_r^0'] + data['DeltaM_r']

# Calculate distance D=Z in parsecs
data['D'] = 10 * 10**((data['r'] - data['M_r'])/5.)
```

# Question 1

What is the volume of the bins in $Z$? The volume of a spherical wedge, bounded radially by concentric spheres is

$$ V = \int_0^{2\pi} d\phi \int_0^{10^\circ} d\theta \int_{R_0}^{R_1} dr \, r^2 \sin\theta $$
$$ = \frac{2\pi}{3} \left[ r^3 \right]_{R_0}^{R_1} \left[ \cos\theta \right]_{10^\circ}^{0} $$


```python
def getdensity(lower, upper, metallicitylimit=None, Nbins=50):
    '''
    Find density of stars in range of Z bins with
    lower < g - r < upper.
    '''
    # Get stars within color limits:
    x = data['g'] - data['r']
    incolorlimits = (lower < x)*(x < upper)
    
    # For question 3:
    # Get stars within metallicity limits:
    if metallicitylimit is None: 
        # If no metallicity limits are specified, make array of Trues
        inmetallicitylimit = np.ones_like(x).astype(bool)
    else:
        inmetallicitylimit = metallicitylimit.copy()

    inalllimits = incolorlimits * inmetallicitylimit # Logical AND
    
    # Compute density histogram for stars in the g-r color range
    N_z, binedges = np.histogram(data['D'][inalllimits], Nbins, range=[0,7e4])
    z_bincenters_pc = [0.5*(binedges[i] + binedges[i+1]) for i in range(len(binedges) - 1)]
    z_bincenters_kpc = 1e-3 * np.array(z_bincenters_pc)
    V_z = [2*np.pi/3 * (binedges[i+1]**3 - binedges[i]**3) * (np.cos(0) - np.cos(np.radians(10))) 
           for i in range(len(binedges) - 1)] 
    rho_z = N_z / V_z

    # For question 1:
    # Compute limit in Z where fractional Poisson errors are <10%
    fractionalerr = 1./np.sqrt(N_z)
    maxreliableZ = np.max(z_bincenters_kpc[fractionalerr <= 0.10])
    
    # For question 4:
    rmags = data['r'][inalllimits]
    
    return z_bincenters_kpc, rho_z, N_z, maxreliableZ, rmags
    
z_bincenters_kpc, rho_z, N_z, maxreliableZ, _ = getdensity(0.2, 0.4)
    
fig, ax = plt.subplots(1, 2, figsize=(16,8))
ax[0].plot(z_bincenters_kpc, np.log(rho_z), ls='steps')
ax[0].set_ylabel(r'$\log \rho(Z) \, \approx \, \log \left( N(Z)/V(Z) \right)$')

ax[1].plot(z_bincenters_kpc, 1./np.sqrt(N_z), 'o')
ax[1].set_ylabel(r'Fractional uncertainty of $ N(Z) = 1/\sqrt{N}$')

for axis in ax:
    axis.set_xlabel('$Z$ [kpc]')
    axis.axvline(maxreliableZ, ls='--', color='k')
    y_annotate = axis.get_ylim()[1] - 0.05*abs(np.diff(axis.get_ylim()))
    axis.annotate('Reliable', xy=(maxreliableZ-2, y_annotate), ha='center', va='top', rotation=90)
    axis.annotate('Unreliable', xy=(maxreliableZ+3, y_annotate), ha='center', va='top', rotation=90)
    ylims = axis.get_ylim()
    axis.fill_between(np.linspace(maxreliableZ, np.max(z_bincenters_kpc)), 
                      ylims[0], ylims[1], color='k', alpha=0.1)
    axis.set_ylim(ylims)

plt.show()
```

Above I've computed the number of stars in 50 bins of $Z$, and computed the approximate log-density of stars with height by dividing the number of stars in a given $Z$ bin by the volume of that bin $Z$. 

To limit ourselves to the data that are most reliable, I've also plotted the fractional uncertainty in the number of stars in a bin $N(Z)$. From Poisson counting statistics, the uncertainty in a number of counts $N$ is $\sqrt{N}$, so 
$$\frac{\sigma_{N(Z)}}{N(Z)} = \frac{1}{\sqrt{N(Z)}}$$

I made the arbitrary decision that 10% fractional error is the upper limit of what I might call a "precise" measurement, and judging by that standard we can reliably measure the density out to roughly **35 kpc** using these data.

## Question 2

```python
# Additional color limits: 
colorlimits = [(0.2, 0.4), (0.4, 0.6), (0.6, 0.8), (0.8, 1.0)]
fig, ax = plt.subplots(1, figsize=(12,8))

for limits in colorlimits: 
    z_bincenters_kpc, rho_z, _, _, _ = getdensity(*limits)
    ax.plot(z_bincenters_kpc, np.log(rho_z), ls='steps', 
            label='${0} < g-r < {1}$'.format(*limits))
ax.set_ylabel(r'$\log \rho(Z)$')
ax.set_xlabel('$Z$ [kpc]')
ax.legend()
plt.show()
```

**Discussion**: The $0.2 < g-r < 0.4$ color bin has higher densities at large heights than the other color bins. Though all color bins have similar densities at small heights, the density of the reddest stars drops off very quickly with height, while the blue stars are relatively more broadly distributed. This is suggestive of the luminosity-$T_{eff}$ relation. Most of the stars are likely on the main sequence since they spend most of their lives there. The luminosities of blue main sequence stars are much greater than the luminosities of red ones, and therefore can be observed to much greater distances in magnitude-limited surveys like SDSS. This is why we can measure the density of the bluest stars out to $\sim$35 kpc while we run out of stars to count in the reddest bin near 10kpc.

We should expect larger systematic errors in the redder bins because the redder stars are much dimmer than blue stars at similar distances, so our uncertainties on the density are relatively larger.

**Conclusion**: The vast majority of metal rich stars in this color bin are typically found at small heights (in the peak for $Z < 5$ kpc), while the metal poor stars can be described reasonably well by one power-law distribution of $\rho(Z)$. 

**Simple models**: The number of metal poor stars in each $r$ bin follows approximately an exponential pattern on $14 < r < 19.5$, while the number of metal poor stars seems to follow a distribution in $r$ magnitudes better described by the sum of two gaussian profiles. This could be indicative of one relatively homogeneous population of metal poor stars, and two distinct populations of metal rich stars. 

## Question 5

```python
# Get r and Z for stars within color range
x = data['g'] - data['r']
incolorlimits = (0.2 < x)*(x < 0.4)
rmags = data['r'][incolorlimits]
logZ = np.log10(1e-3*data['D'][incolorlimits])

# Extrapolate to 100 kpc
pfit = np.polyfit(rmags, logZ, 1)
faintlimit = (2 - pfit[1])/pfit[0]
x_model = np.linspace(14, faintlimit, 10)
y_model = np.polyval(pfit, x_model)

fig, ax = plt.subplots(1, figsize=(8,8))
ax.plot(rmags, logZ, '.', alpha=0.02, label='SDSS')
ax.plot(x_model, y_model, 'r', lw=2, label='Trend')
ax.axhline(2, ls='--', color='k')
ax.axvline(faintlimit, ls='--', color='k')

ax.annotate('100 kpc', xy=(16, 2), va='bottom')
ax.annotate('Faint limit:\n $r$ = {0:.2f}'.format(faintlimit), 
            xy=(22, 0.1), ha='center')

ax.set_xlabel('$r$')
ax.set_ylabel('$\log\,Z$ [kpc]')
ax.set_ylim([-0.5, 2.5])
plt.show()
```

For a survey to measure the density of stars out to 100 kpc, it will need a deeper faint limit than SDSS. First I transform into $(r, \log\,Z)$ space since $r$ is already logarithmic, allowing use to model the correlation with a simple linear function. 

Extrapolating the correlation between the log of the galactic height $Z$ and $r$ magnitude, typical main sequence stars at $Z=100$ kpc should have $r \approx 24.7$, so the faint limit for a survey that probes $\log\,\rho(Z)$ near 100 kpc should have a **faint limit near $r=25$**. 

As a sanity check, I found that the [LSST](http://lsst.org/lsst/science/scientist_milky_way) intends to obtain metallicity measurements out to 100 kpc with its $r$ faint limit of 24, so my very rough estimate of $r=25$ looks reasonable.

### Number of Stars at 90-100 kpc

In order to estimate the number of stars there are with $0.2 < r < 0.4$ between 90 and 100 kpc in a 1 deg$^2$ field of view, we can extrapolate from the number of stars in each distance bin calculated in Question 1-2. I'll only use the data within the "reliable" distance limit of 35 kpc.

```python
z_bincenters_kpc, rho_z, N_z, maxreliableZ, rmags = getdensity(0.2, 0.4)

reliablerange = (z_bincenters_kpc > 10)*(z_bincenters_kpc < maxreliableZ)

pfit = np.polyfit(z_bincenters_kpc[reliablerange], 
                  np.log10(N_z[reliablerange]), 1)

x_model = np.arange(0,101)
y_model = np.polyval(pfit, x_model)

# How many stars in this color range should you expect to find 
# out between 90-100 kpc?
Ninrange = np.trapz(10**y_model[x_model > 90], x_model[x_model > 90])
print('Stars with 0.2 < r < 0.4 at 90-100 kpc: {0:.3f}'.format(Ninrange))

fig, ax = plt.subplots(1, figsize=(12,8))

ax.plot(z_bincenters_kpc, np.log10(N_z), ls='steps')
ax.plot(x_model, y_model, 'r', lw=2, label='Trend')

ax.set_ylabel(r'$\log N$')
ax.set_xlabel('$Z$ [kpc]')
ax.set_xlim([0,100])
plt.show()
```

We need to extend the trend in the number of stars with $0.2 < g-r < 0.4$ within some distance range, out to $90 < Z < 100$ kpc. Since the overall exponential decline in the number of stars with distance only holds from about $Z\approx 10$ kpc to the "maximum reliable $Z$" ($Z\sim35$ kpc), I choose this distance range as my training set for what to expect out at higher $Z$ where our statistics are limited by small numbers (or no counts). I found that you'd expect $N < 1$ stars to be found at that distance.

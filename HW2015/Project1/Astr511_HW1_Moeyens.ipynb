```python
import gzip
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii 
from IPython.display import display

plt.rc('axes', labelsize=16)
plt.rc('xtick', labelsize=12) 
plt.rc('ytick', labelsize=12)
plt.rc('legend', fontsize=12)
plt.rc('font', size=18)

% matplotlib inline
```

```python
# Unzip data and read
f = gzip.open('Astr511HW1data.dat.gz')
data = ascii.read(f)
f.close()
```

```python
# Define relevant functions

def metallicity(P, ug, gr):
    """Returns metalicity given an array of best-fit coefficients and colors: u-g, g-r."""
    # Equation from Bond et al. 2010 (ApJ, 716, 1)
    met = np.ones(len(ug))
    for i,j in enumerate(gr):
        if j < 0.6:
            met[i] = (P[0] + P[1]*ug[i] + P[2]*j + P[3]*ug[i]*j + P[4]*(ug[i]**2) + P[5]*(j**2) 
                 + P[6]*(ug[i]**2)*j + P[7]*ug[i]*(j**2) + P[8]*(ug[i]**3) + P[9]*(j**3))
        else:
            met[i] = -0.6
    
    lim1 = np.where(met < -2.5)
    lim2 = np.where(met > 0.5)
    met[lim1] = -2.5
    met[lim2] = 0.5
    
    return met

def M_r0(P, gi):
    """Returns initial absolute magnitude in the r band for given color g-i."""
    # Equation A7 from Ivezic et al. 2008 (ApJ, 684, 287)
    return (P[0] + P[1]*gi + P[2]*(gi**2) + P[3]*(gi**3) + P[4]*(gi**4) + P[5]*(gi**5))

def dM_r(P, met):
    """Returns the change in absolute r band magnitude given an array of best-fit coefficients and metallicity."""
    # Equation A2 from Ivezic et al. 2008 (ApJ, 684, 287)
    return (P[0] + P[1]*met + P[2]*(met**2))

def M_r(M_r0, dM_r):
    """Returns the absolute r band magnitude given an initial r band magnitude and the change in r band."""
    # Equation A2 from Ivezic et al. 2008 (ApJ, 684, 287)
    return M_r0 + dM_r

def distance(r, Mr):
    """Returns distance in kpc given r band magnitudes and absolute magnitudes. """
    return 10.0 * (10.0**((r - Mr)/5.0))

def binVolume(r_in, r_out):
    """Returns volume of spherical wedge between r_in and r_out. """
    return (4.0/3.0) * np.pi * (r_out**3.0 - r_in**3.0) * (314.0/41253.0)
 
def lnStellarDensity(Z, color, colorRange, m=None, mLow=None, mHigh=None, bins=20):
    """Returns the natural log of the stellar density, the histogram distribution of stars, the bin volume and edges."""
    
    if mHigh != None:
        z = Z[(color > colorRange[0])*(color < colorRange[1])*(m > mHigh)]
    elif mLow != None:
        z = Z[(color > colorRange[0])*(color < colorRange[1])*(m < mLow)]
    else:
        z = Z[(color > colorRange[0])*(color < colorRange[1])]   
        
    N = len(z)
    cumSum = np.cumsum(z)
    sigma_g = 0.7413*(0.75*cumSum - 0.25*cumSum)
    delta_bins = ((2.7*sigma_g) / N**(0.25))
    
    # bins = (z.max()-z.min()) / delta_bins
    
    #print bins
    #print len(bins)
    
    hist, bin_edges = np.histogram(z,bins=bins)
    bin_vol = binVolume(bin_edges[:-1], bin_edges[1:])
    ln_Rho = np.log(hist / bin_vol)
    return ln_Rho, hist, bin_vol, bin_edges    
```

```python
# Initialize parameter arrays
P = [-13.13,14.09,28.04,-5.51,-5.90,-58.68,9.14,-20.61,0.0,58.20]
A7 = [-5.06,14.32,-12.97,6.127,-1.267,0.0967]
A2 = [4.50,-1.11,-0.18]

# Prepare relevant data, note: resulting column headers will be wrong
r_n = data['r']    # limit sample to r < 21.5
rlimit = np.where(r_n < 21.0)
g_r = np.array(data['g'][rlimit] - data['r'][rlimit])
g_i = np.array(data['g'][rlimit] - data['i'][rlimit])
u_g = np.array(data['u'][rlimit] - data['g'][rlimit])
r = np.array(data['r'][rlimit])

# Calculate Z
m = metallicity(P, u_g, g_r)
Mr0 = M_r0(A7, g_i)
dMr = dM_r(A2, m)
Mr = M_r(Mr0, dMr)
Z = distance(r, Mr)
```

### Question 1:

Plot ln(rho) vs Z for a stellar color subset of 0.2 < g-r < 0.4. What is the Z range where you believe the results?

```python
ln, hist, vol, edges = lnStellarDensity(Z,g_r,[0.2,0.4])

fig, ax = plt.subplots(1,1)
fig.tight_layout()
fig.suptitle('ln(' + r'$\rho$' + ') vs. Z',fontsize=14,y=1.05)

ax.step(edges[1:]/1000.0,ln);
ax.hlines(xmin=0,xmax=edges[1]/1000.0,y=ln[0],color='blue')

major_ticksX = np.arange(0, 100, 20)                                              
minor_ticksX = np.arange(0, 100, 5) 
major_ticksY = np.arange(-30,-10,5)
minor_ticksY = np.arange(-30,-10,1)

ax.set_xticks(major_ticksX)                                                       
ax.set_xticks(minor_ticksX, minor=True)                                           
ax.set_yticks(major_ticksY)
ax.set_yticks(minor_ticksY, minor=True)
ax.grid()
ax.axvline(10,color='gray',linestyle='--',label='sanity limit')
ax.legend()

ax.set_xlabel(r'$Z$ (kpc)');
ax.set_ylabel(r'$ln(\rho$)');
```

The range of believable results should be between 0 and 10 kpc. At the upper limit r = 21 for the SDSS sample which limits r - M_r to approximately 15 magnitudes. The distance modulus at 15 magnitude difference approximates to about 10 kpc. 

### Question 2:

Plot ln(rho) vs Z for different subsets of g-r color. Why do we expect larger systematic errors for 0.8 < g-r < 1.0 than the adjacent bin with 0.4 < g-r < 0.6?


```python
ln1, hist1, vol1, edges1 = lnStellarDensity(Z,g_r,[0.2,0.4])
ln2, hist2, vol2, edges2 = lnStellarDensity(Z,g_r,[0.4,0.6])
ln3, hist3, vol3, edges3 = lnStellarDensity(Z,g_r,[0.6,0.8])
ln4, hist4, vol4, edges4 = lnStellarDensity(Z,g_r,[0.8,1.0])

lns = [ln1,ln2,ln3,ln4]
hists = [hist1,hist2,hist3,hist4]
vols = [vol1,vol2,vol3,vol4]
edgesList = [edges1,edges2,edges3,edges4]
colors = ['blue','green','orange','red']
names = [r'$0.2 < g-r < 0.4$',r'$0.4 < g-r < 0.6$',r'$0.6 < g-r < 0.8$',r'$0.8 < g-r < 1.0$']

fig, ax = plt.subplots(1,2)
fig.set_size_inches(10,4)
fig.tight_layout()
fig.suptitle('ln(' + r'$\rho$' + ') vs. Z for Different Color Populations',fontsize=14,y=1.05)
plt.subplots_adjust(wspace=0.2)

for i,j in enumerate(lns):
    ax[0].step(edgesList[i][1:]/1000.0,j,color=colors[i]);
    ax[0].hlines(xmin=0,xmax=edgesList[i][1]/1000.0,y=lns[i][0],color=colors[i],label=names[i])
    ax[1].step(edgesList[i][1:]/1000.0,j,color=colors[i]);
    ax[1].hlines(xmin=0,xmax=edgesList[i][1]/1000.0,y=lns[i][0],color=colors[i],label=names[i])
    
major_ticksX = np.arange(0,100,20)          
minor_ticksX = np.arange(0,100,5)
major_ticksY = np.arange(-30,0,5)
minor_ticksY = np.arange(-30,0,1)

for i in ax:
    i.set_xticks(major_ticksX);                                                  
    i.set_xticks(minor_ticksX, minor=True);                                          
    i.set_yticks(major_ticksY);
    i.set_yticks(minor_ticksY, minor=True);
    i.grid()
    i.set_xlabel(r'$Z$ (kpc)');
    i.set_ylabel(r'$ln(\rho$)');
    
ax[1].set_xticks(np.arange(0,20,5));
ax[1].set_xticks(np.arange(0,20,1),minor=True);
ax[1].set_xlim(0,10);
ax[0].legend();
```

The plots above clearly show that the redder the population of stars the faster the stellar density decreases as Z gets larger. The second plot indicates that at very low Z (Z < 2.5 kpc) the population density of redder stars is greater as redness increases. A star's redness is typically closely associated to its age, this would suggest that at very low Z or towards the galactic center there is a large population of older stars and that on average age of stars decreases as Z increases. 

We would expect larger systematic errors for redder stars as they are dimmer than blue stars when looking at the same distance. The dimmer an object is the more the systematic errors from hardware tolerances (mirrors, filters, etc... ) can have an effect on the measured values. In relation to the above, we would also expect fewer and fewer red star detections the further out we go as they become too faint to see.

The plot above shows that low metallicity stars are more frequent than high metallicity stars at lower Z. Beyond about  20 kpc high metallicity stars are more frequent than low metallicity stars. Typically, metallicity is a good indicator of a star's age: younger stars have higher metallicity than an older and lower metallicity stars. This shows that there is a large population of older stars at low Z (Z < 5kpc) with a significant population of younger stars between 0 < Z < ~1.2. As Z increases the amount of older stars decreases more rapidly than their younger counterpart.

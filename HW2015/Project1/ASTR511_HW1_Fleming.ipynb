#Define functions used to calculate M_r, rho(z)
The below functions are used to calculate the parameters Delta M_r, M_0, and the metallicity [Fe/H].  These parameters are then used in an additional function to calculate that absolute r magnitude, M_r.  Also included is a function that given a subset of data to look for, computes the stellar density as a function of distance z.

To compute $M_r$, equations A2, A3 and A7 from Ivezic et al. 2008 (ApJ,684,287) are employed.

$$A2: \Delta M_r([Fe/H]) = 4.50-1.11[Fe/H]-0.18[Fe/H]^2$$
$$A3: M_r(g-i,[Fe/H]) = M^0_r(g-i) + \Delta M_r([Fe/H])$$
$$A7: M^0_r(g-i)=-5.06+14.32x-12.97x^2+6.127x^3-1.267x^4+0.0967x^5 \ \ \ \ where \ \ x = g-i $$

From Bond et al. 2010 (ApJ, 716, 1), the metallicty [Fe/H] is given by: 

$$[Fe/H] = A + Bx + Cy + Dxy + Ex^2 + F y^2 + Gx^2y + Hxy^2 + Ix^3 + Jy^3$$ 

where x = (u − g) and y = (g − r).  The best-fit parameters (A–J) = (−13.13, 14.09,
28.04, −5.51, −5.90, −58.68, 9.14, −20.61, 0.0, 58.20).  Note: this expression if valid only for the case when
g − r < 0.6.  For g - r >= 0.6, [Fe/H] = −0.6

These equations are all given in the functions defined below and used to calculate various parameters of interest such as as the r band absolute magnitude $M_r$.

```python
#Define functions to use later
def calcDeltaMR(metals):
    """
    Function takes in Fe/H metalicity, outputs Delta_M_r using A2 from Ivezic+2008
    """
    return (4.50-1.11*(metals)-0.18*(metals*metals))
#end function

def calcMr(M_0,deltaMr):
    """
    Function takes Fe/H metalicity, g-i color, and M_0, computes absolute r magnitude Mr
    """
    return (M_0 + deltaMr)
#end function

def calcM_0(gi):
    """
    Function takes g-i color, calculates M_0 parameters
    """
    return (-5.06 + 14.32*gi - 12.97*(gi*gi) + 6.127*(gi*gi*gi) - 1.267*(gi*gi*gi*gi) + 0.0967*(gi*gi*gi*gi*gi))

def calcFeH(x,y):
    """
    Function computes Fe/H metalicity given x (u-g) and y (g-r) colors and index i
    Can't accept arrays since need to do logical comparison to see if it's in valid range
    """
    A = -13.13
    B = 14.09
    C = 28.04
    D = -5.51
    E = -5.90
    F = -58.68
    G = 9.14
    H = -20.61
    I = 0.0
    J = 58.20
    
    #Valid only for g-r < 0.6
    if(y >= 0.6):
        return -0.6
    else:
        return (A + B*x + C*y + D*x*y + E*x*x + F*y*y + G*x*x*y + H*x*y*y + I*x*x*x + J*y*y*y)
    
def calcRho(zArray,mask,bins=100,omega=314):
    """
    Given a data mask, the array of stellar distances, bin constant and solid angle,
    return density and distance arrays for plotting
    
    Input:
    zArray: Array of stellar distances Z
    mask: Mask for properly selecting data of interest
    bins: Number of bins for histogram
    omega: solid angle
    
    Output:
    rho: rho(Z) array of density as a function of distance
    zBins: array of distances for plotting
    """
    #Selects distance bins that fit the color criteria
    zPrime = zArray[mask]
    
    #Bin stars appropriately
    N, edges = np.histogram(zPrime,bins)
    zBins = (edges[1:] + edges[:-1])/2.0
    
    binSize = (np.power(edges[1:],3.0) - np.power(edges[:-1],3.0))/np.power(1000.0,3)
    
    #Convert zBins from pc to kpc
    zBins = zBins/1000.0
    
    #Compute volume of each bins
    scale = np.pi/(180.0)
    V = omega*scale*binSize/3.0
    
    #Compute density
    rho = N/V
    
    return rho, zBins
```

Define a function to bin data depending on what magnitude band you're considering and what subset of data you wish to study.  The results of this function can be used to make a differential magnitude distribution for a given band.  The volume of each bin is given by $V = \Omega(r^3_2 - r^3_1)/3$ where $\Omega = sin(\theta)d \theta d \phi$ is the solid angle.

```python
#Define a function to histrogram stars by magnitude (for question 4)
def magBin(data=None,band=6,mask=None,size=0.1):
    """
    Given the data object, a mask to select a subset of the data, the magnitude key word and bin size, 
    return arrays to produce a scatter plot of the differential (x) band magnitude distribution for that sample.
    
    Inputs:
    Data: data array for all objects.  Assumes it's of the form data[number of objects,band]
    band: which band to select for.  Must be as an int flag like R=6 for the r band given below
    mask: allows to select for a certain data subset from the Data array (optional)
    size: Width of bins
    
    Outputs:
    values, rBins to produce a plot of the form scatter(rBins,values)
    """
    #Create subset if mask given
    if mask != None:
        subset = data[mask]
    else:
        subset = data
    
    #Select correct band
    subset = subset[:,band]
        
    #Compute number of bins for histogram
    bins = (subset.max()-subset.min())/size
    
    #Create histogram
    vals, edges = np.histogram(subset,bins)
    rBins = (edges[1:] + edges[:-1])/2.0
    
    #Return results
    return rBins, vals
```

#What is the Z range where you believe the results, and why?
This figure, when compared to Fig. 5 of Juric et al 2008, demonstrates that the galaxy's stellar number density profile is a composite of two power law components.  The inner power law which approximately runs from Z = 0 to Z ~ 4 kpc (dashed black line) corresponds to the galactic disk while the power law component for Z > 4 kpc corresponds to the galactic halo (Juric 2008, Fig. 5 caption).  Given this information, one would imagine that the Z range where the results become tenuous would be in the far out galactic halo since the light from those objects would be faint due to the distance it has to travel and the intervening media attenuating it.  In order to calculate the Z range where one would believe the results, one can make the following argument.  If one assumes that halo stars have some $M_r$ distribution with some mean value and standard deviation and that the spectroscopic data for those stars is accurate, the mean $M_r$ and corresponding standard deviation for the halo population can be calcuated given $g-i$ and $[Fe/H]$ using the formulas above.  I'll assume that this $M_r$ sufficiently represents the general properties of the halo such that far out stars and nearer stars in the halo will be within 1 sigma of this value in general.

Note that $M_r$ is a function of the metallicity $[Fe/H]$ and the $g-i$ color.  After computing $M_r$, one can derive Z from $r − M_r = 5 ∗ log (D/(10pc))$ under the approximation that Z ~ D.  To probe the outer limit at which we believe this data, I'll select the limiting r band magnitude, 21.

Note: Instead of the mean values for the above quantities, the median values are chosen in order to attempt to remove the effects of outliers.

```python
#Compute the median [Fe/H] of the Z > 4 kpc population stars
mask = np.logical_and(zArray > 4000,np.logical_and(0.2 < (data[:,G] - data[:,R]),(data[:,G] - data[:,R]) < 0.4))

#Select correct data, find median
haloMetals = fehArray[mask]
haloMetals_median = np.median(haloMetals)
haloMetals_std = np.std(haloMetals)
print "Median [Fe/H]: {0}, standard deviation: {1}.".format(haloMetals_median, haloMetals_std)
```

With the median halo metallicity given above, we simply need the median g-i color of this sample before we can calculate $M_r$ and hence the distance Z.  Similar to the above cell, I'll compute the median g-i color for the given sample and then use that to calculate $M_r$ and the corresponding distance Z.

```python
#Compute median g-i color for Z > 4 kpc population stars where 0.2 < g-r < 0.4

#Compute g-i for correct stars
gi_Z = data[:,G] - data[:,I]
gi_Z = gi_Z[mask]

#Find median
gi_median = np.median(gi_Z)
gi_std = np.std(gi_Z)
print "Median g-i: {0}, standard deviation: {1}.".format(gi_median, gi_std)
```

Using these values, we can now compute $M_r$ and then Z in kpc.  To calcualte distance via the distance modulus, we need one final value: the limiting r band magnitude, 21.  The calculations are given below.

```python
#Compute absolute magnitude then distance
M_r_halo = calcMr(calcM_0(gi_median),calcDeltaMR(haloMetals_median))
M_r_halo_std = calcMr(calcM_0(gi_std),calcDeltaMR(haloMetals_std))
distance = (10.0)*np.power(10.0,((21.0-M_r_halo)/5.0))/(1000.0)

print "Approximate maximum distance where which we believe our results: {0} kpc.".format(distance)
```

By calculating the median $M_r$ for the Z > 4 kpc halo population and assuming the faint r magnitude limit of 21, one can approximate the maximum distance at which we believe the results to be around 18 kpc.

#Discuss the differences compared to the 0.2 < g − r < 0.4 subsample.

As described above, each curve consists of two power law components, one for the disk for Z < 4 kpc and one for the halo beyond that point.  The disk power law component for each color bin all seem to be roughly the same implying the stellar number density in the disk has a weak dependence on g-r color at best.  In the halo as color increases from 0.2 < g-r < 0.4 to 0.8 < g-r < 1, however, each curve's halo power law component becomes steeper indicating that stellar number density drops dramatically with increasing color.

The physical nature of this stems from the g-r color.  A large g-r color indicates that the population consists of redder, older stars relative to neighboring bins.  Therefore the steeper the halo power law component, the redder and likely older the population.  From this, one may conclude that since on average red, old main sequence stars are dimmer and since red light suffers from ISM extinction more than bluer light, redder populations would be more difficult to detect as one goes out to higher Z, explaining the steeper power laws and limited Z extent we see above.

#Why do we expect larger systematic errors for 0.8 < g − r < 1.0 than for the adjacent bin with 0.4 < g − r < 0.6?

We expect larger systematic errors for 0.8 < g-r < 1.0 than for the adjacent 0.4 < g-r < 0.6 bin because as discussed above, the former bin consists generally of redder, older main sequence stars.  Since old, red stars tend to be dimmer than their blue counterparts and due to ISM extinction more strongly attenuating red light, 0.8 < g-r < 1.0 stars will be much more difficult to detect and suffer from selection bias.  This effect can be seen in the steeper halo power law component as g-r color increases.

#Question 5: What should be the faint r band limit for a survey to be able to map the ln(ρ) vs. Z profile out to 100 kpc using main-sequence stars?

At distances of D = 100 kpc, the distance modulus is $20 = 5\times log_{10}(D/10pc)$.  If we can determine roughly what $M_r$ is, then we can infer r given that the distance modulus is $r-M_r$.  To calculate $M_r$, we can assume that main sequence stars at large Z out in the halo will have roughly the same distribution of properties in general where the halo is defined as the power law region for Z > 4 kpc.  Assuming a similar distribution amongst halo stars of all color, the population on average will have some median abosolute magnitude.  The approximate is likely accurate to first order as the median $M_r$ found amongst all color bins for halo stars will sufficiently represent the population.  Under this assumption after computing the median g-i color and median [Fe/H] values, we can find $M_r$ and hence the r limit at Z = 100 kpc.

```python
#Select halo stars where Z > 4kpc where zArray is in units of pc
mask = (zArray > 4000) 

#Select correct g-i color range
giHalo = (data[:,G] - data[:,I])
giHalo = giHalo[mask]

#select correct metallicity range
fehHalo = fehArray[mask]

#Compute median g-i
giHaloMedian = np.median(giHalo)

#Compute median metallicity
fehHaloMedian = np.median(fehHalo)

#Compute M_r
M_rHaloMedian = calcMr(calcM_0(giHaloMedian),calcDeltaMR(fehHaloMedian))

print M_rHaloMedian
```

Given that the estimate for the median $M_r \approx 5.2$ taken over all color bins, we get r = $M_r + 20$ or an r magnitude of about $r = 25.2$ which is beyond the r = 21 SDSS limit as expected.  Therefore if a survey wishes to map the $ln(\rho) \ vs \ Z$ distribution out to around 100 kpc, it must be sensative enough to detect r band magnitudes of about 25.

#For a solid angle of 1 deg^2, how many starswith 0.2 < g − r < 0.4 would you expect with distances between 90 kpc and 100 kpc?
For this question, assume $\rho \sim \rho_0r^{p}$  where $\rho_0$ and the power be will be found via a fit.  To fit the data, I will use numpy's polyfit function to fit a linear function for distance Z vs. log($\rho$).  I will fit the data over halo stars, Z > 4 kpc, since distances of ~ 90-100 kpc correspond to the farthest out halo stars.  To confirm the validity of the fit, I will plot the fitted values over the true values in order to roughly validate the correspondance.  Once the fit gives me a density estimate, I can then calculate the volume in the 90-100 kpc bin and then use those to values to come up with an estimate for the number of stars.

```python
mask = np.logical_and(zBins > 4, zBins < 40)

#rho, zBins correspond to the 0.2 < g-r < 0.4 data used in question 1.  Select the halo population from these stars, perform fit
rhoHalo = rho[mask]
zHalo = zBins[mask]

#Perform fit
m, b = np.polyfit(zHalo,np.log(rhoHalo),1)

#Define a function to use for transforming coordinates/plotting
def f(m,b,x):
    return np.exp(m*x+b)

#Compute volume element where omega = 1 degree
omega = 1.0
binSize = ((100.0*100.0*100.0) - (90.0*90.0*90.0))/3.0
scale = np.pi/(180.0)
V = omega*scale*binSize
```

```python
#Produce comparison plot

#Plot actual data
plt.plot(zBins,np.log(rho),linestyle='steps-mid',linewidth=2,label='True Values')

#Create fit points to plot
x = np.linspace(0,110,1000)
y = f(m,b,x)

#Plot fit results
plt.plot(x,np.log(y),label='Fit')

#Style plot
plt.xlim([0,110])
plt.ylim([-24,8])
plt.xlabel("$Z \ (kpc)$")
plt.ylabel(r"$log(\rho(Z))$")
plt.title(r"$log(\rho(Z)) \ vs. \ Z \ for \  0.2 \ < \ g-r \ < \ 0.4$")
plt.legend()
plt.show()
```

It appears as if the fit for the halo power law component is accurate for most values of Z, but note that beyond Z ~ 40 kpc, the fit deviates.  This deviation likely corresponds to an accumulation of systematic error in the data such that the data cannot be trusted at these distances.  One useful note about this plot is that it also clearly illustrates the difference between the halo power law component and that of the Z < 4 kpc disk.  Now with the estimated density and volume, compute the number of stars.

```python
print "Estimated number of stars in the middle of the Z = [90,100] kpc bin = {0}".format(f(m,b,95.0)*V)
```

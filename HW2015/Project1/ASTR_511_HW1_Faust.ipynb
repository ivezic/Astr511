## Assignment 1: Tomography of MWG by high-galactic-azimuth SDSS stars
### Get and clean data from SDSS star data file

```python
# Open files and clean data
import numpy as np
starfile = open("Astr511HW1data.dat", "r")
head = starfile.readlines()    
starfile.close()
N = len(head)

Star_Mags = np.ones(N, dtype={'names':['u','g','r','i','z','FeH','Mr','Z'], 'formats':['f8','f8','f8','f8','f8','f8','f8','f8']})

for x in range(N-2):  
        Star_Mags['u'][x] = float(head[x+2][35:42]) # best fit u
        Star_Mags['g'][x] = float(head[x+2][43:48]) # best fit g 
        Star_Mags['r'][x] = float(head[x+2][49:55]) # best fit r 
        Star_Mags['i'][x] = float(head[x+2][56:62]) # best fit i 
        Star_Mags['z'][x] = float(head[x+2][63:70]) # best fit z  
        
print(Star_Mags[0:1])
```

### Use magnitudes to compute metallicity (FeH), Absolute r-band Magnitude (Mr) and Distance above galactic plane (Z) \approx Distance above Sun  

#### Photometric metallicity as in Bond et al. 2010 (ApJ, 716, 1)
#### Photometric parallax as in Ivezic et al. 2008 (ApJ, 684, 287) appendix A

```python
# compute metallicity and abs mag_r via photometric parallax calculation and assumption that all are MS
# function to compute photometric metallicity

def Metallicity(u, g, r):
    y = g - r
    if(y >= 0.6): 
        return -0.6
    else:
        x = u - g
        M = -13.13 + 14.09*x + 28.04*y - 5.51*x*y - 5.9*x**2 - 58.68*y**2 + 9.14*x**2*y - 20.61*x*y**2 + 0.0*x**3 + 58.20*y**3
        return M

# function to compute absolute magnitude via photometric parallax
def AbsoluteMagnitude(g, i, FeH):
    #Following naming conventions in Ivezic et al appendix A
    DeltaMr = 4.50 - 1.11*FeH - 0.18*FeH**2
    Mr0 = -2.85 + 6.29*(g-i) - 2.30*(g-i)**2
    return DeltaMr+Mr0

# find height above galactic plane -that is us- in kPc 
# from distance-magnitude relation and approximation that distance and height above galactic plane are equal
def Z_galactic_plane(r, Mr):
    return 10**(0.2*(r - Mr)-2)


for x in range(N):
    Star_Mags['FeH'][x] = Metallicity(Star_Mags['u'][x], Star_Mags['g'][x], Star_Mags['r'][x]) 
    
for x in range(N):
    Star_Mags['Mr'][x] = AbsoluteMagnitude(Star_Mags['g'][x], Star_Mags['i'][x], Star_Mags['FeH'][x])
 
for x in range(N):
    Star_Mags['Z'][x] = Z_galactic_plane(Star_Mags['r'][x], Star_Mags['Mr'][x])
    
print(Star_Mags[0:1])
```

### First two questions.  Make color-color cuts on (g-r) and look at populations vs. height about galactic plane.

```python
# question 1 select stars with 0.2 < g-r < 0.4
gr_band1 = Star_Mags[np.logical_and((Star_Mags['g']-Star_Mags['r']) > 0.2, (Star_Mags['g']-Star_Mags['r']) < 0.4)]
# slices for question 2
gr_band2 = Star_Mags[np.logical_and((Star_Mags['g']-Star_Mags['r']) > 0.4, (Star_Mags['g']-Star_Mags['r']) < 0.6)]
gr_band3 = Star_Mags[np.logical_and((Star_Mags['g']-Star_Mags['r']) > 0.6, (Star_Mags['g']-Star_Mags['r']) < 0.8)]
gr_band4 = Star_Mags[np.logical_and((Star_Mags['g']-Star_Mags['r']) > 0.8, (Star_Mags['g']-Star_Mags['r']) < 1.0)]
band_counts = [len(gr_band1), len(gr_band2), len(gr_band3), len(gr_band4)]
print(band_counts)

# program to bin data manually and process into ln densities.  
import math 

def distance_bin(StarNFO, Dens, Dmax, n, nBins):
    index = -1.0
    err_cnt = 0.0
    for x in range(nBins):
        Dens[x] = 0
    
    for x in range(n):
        if(StarNFO['Z'][x] > Dmax):
            err_cnt += 1
        if(StarNFO['Z'][x] < Dmax):
            index = math.floor((bins*StarNFO['Z'][x])/Dmax)     
            Dens[(int(index))] += 1
    message = "finished binning, " + str(err_cnt) + " stars out of range"
    print(message)
    # transform count to ln of physical density
    SphFrac = (314.0/41235.0)*(4.0*3.141593/3.0)
    for y in range(nBins):
        Bin_Vol = SphFrac*((Dmax*float(y+1)/float(nBins))**3-(Dmax*float(y)/float(nBins))**3)
        if (Dens[y] == 0):
            Dens[y] += 1  # fake count to keep zero-count bins from crashing
  #      print(Dens[y])
        Dens[y] = math.log(Dens[y]/Bin_Vol)
    return(Dens)

bins = 15
Dmax = 40 # max plot distance in kPc
StarDens1 = np.zeros((bins))# StarDens1 will be log density per volume slice for band 1 etc.
StarDens2 = np.zeros((bins))
StarDens3 = np.zeros((bins))
StarDens4 = np.zeros((bins)) 

distance_bin(gr_band1, StarDens1, Dmax, band_counts[0], bins)
distance_bin(gr_band2, StarDens2, Dmax, band_counts[1], bins)
distance_bin(gr_band3, StarDens3, Dmax, band_counts[2], bins)
distance_bin(gr_band4, StarDens4, Dmax, band_counts[3], bins)
```

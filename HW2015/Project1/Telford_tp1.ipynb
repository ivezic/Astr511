```python
# import relevant packages and set up plots
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rc('xtick', labelsize=16)     
mpl.rc('ytick', labelsize=16)  
mpl.rc('axes', labelsize=20)
mpl.rc('legend', fontsize=20)
mpl.rc('lines', linewidth=2)
mpl.rc('font', size=18, family='serif', style='normal', variant='normal', stretch='normal', weight='normal')
%matplotlib inline
```

```python
# load the data
data = np.loadtxt('Astr511HW1data.dat')
col = np.array(['ra','dec','run','Ar','u','g','r','i','z','uErr','gErr','rErr','iErr','zErr','pmL','pmB','pmErr'])
np.shape(data)
```

```python
# calculate metallicity
ug = (data[:,col=='u'] - data[:,col=='g'])[:,0]
gr = (data[:,col=='g'] - data[:,col=='r'])[:,0]

metallicity = np.ones(np.shape(data)[0]) * -0.6
x = ug[gr < 0.6]
y = gr[gr < 0.6]
metallicity[gr < 0.6] = -13.13 + 14.09 * x + 28.04 * y - 5.51 * x * y - 5.90 * x**2 - 58.68 * y**2 + 9.14 * x**2 * y \
    -20.61 * x * y**2 + 58.20 * y**3
```

```python
# calulate M_r
gi = (data[:,col=='g'] - data[:,col=='i'])[:,0]

delMr = 4.50 - 1.11 * metallicity - 0.18 * metallicity**2
Mr0 = -5.06 + 14.32 * gi - 12.97 * gi**2 + 6.127 * gi**3 - 1.267 * gi**4 + 0.0967 * gi**5
Mr = Mr0 + delMr
```

```python
# calulcate Z in parsecs
Z = 10 * 10**(0.2 * (data[:,col=='r'][:,0] - Mr))
```

## Problem 1: Plot $\ln \rho$ vs. $Z$ for stars with $0.2 < g-r < 0.4$

```python
# define function to calculate number of stars and volume per bin
# volume of spherical ring = volume of outer sphere - volume of inner sphere
# multiply by solid angle divided by total solid angle of sphere: 314 deg^2 / 41253 deg^2

def get_ln_rho(Z, low_color, high_color, bins=50, upperlim=None, low_metallicity=None, high_metallicity=None):
    if low_metallicity:
        Z_samp = Z[(gr > low_color) * (gr < high_color) * (metallicity > low_metallicity)]
    elif high_metallicity:
        Z_samp = Z[(gr > low_color) * (gr < high_color) * (metallicity < high_metallicity)]
    else:
        Z_samp = Z[(gr > low_color) * (gr < high_color)]
    hist, bin_edges = np.histogram(Z_samp, bins=bins)
    vol = (4. * np.pi / 3.) * (314. / 41253.) * (bin_edges[1:]**3 - bin_edges[:-1]**3)
    ln_rho = np.log(hist / vol)

    return bin_edges, ln_rho


# Problem 1 
bin_edges, ln_rho = get_ln_rho(Z, 0.2, 0.4)

fig, ax = plt.subplots(1, 2, figsize=(16,6))
ax[0].set_xlabel(r'$Z$')
ax[0].set_ylabel(r'$\ln \rho$')
ax[0].plot(bin_edges[1:], ln_rho, linestyle='steps', label=r'$0.2 < g-r < 0.4$')

ax[1].set_xlabel(r'$Z$')
ax[1].set_ylabel(r'$\ln \rho$')
ax[1].plot(bin_edges[1:], ln_rho, linestyle='steps', label=r'$0.2 < g-r < 0.4$')
ax[1].plot([0, 65000], [-11.5,-30], 'k--')
ax[1].plot([35000, 35000], [-30, -21.5], 'k--')
```

I believe the results from $Z=0$ pc to roughly $Z=3.5 \times 10^4$ pc, where the histogram begins to deviate from exponential decline.

## Problem 2: Compare $\ln \rho$ vs. $Z$ for stars in different color bins

```python
#Problem 2 
fig, ax = plt.subplots(figsize=(10,8))
ax.set_xlabel(r'$Z$')
ax.set_ylabel(r'$\ln \rho$')

bin_edges, ln_rho = get_ln_rho(Z, 0.2, 0.4, bins=50)
ax.plot(bin_edges[1:], ln_rho, linestyle='steps', label=r'$0.2 < g-r < 0.4$')

bin_edges, ln_rho = get_ln_rho(Z, 0.4, 0.6, bins=30)
ax.plot(bin_edges[1:], ln_rho, linestyle='steps', label=r'$0.4 < g-r < 0.6$')

bin_edges, ln_rho = get_ln_rho(Z, 0.6, 0.8, bins=10)
ax.plot(bin_edges[1:], ln_rho, linestyle='steps', label=r'$0.6 < g-r < 0.8$')

bin_edges, ln_rho = get_ln_rho(Z, 0.8, 1.0, bins=10)
ax.plot(bin_edges[1:], ln_rho, linestyle='steps', label=r'$0.8 < g-r < 1.0$')

plt.legend()
```

The stellar density falls off more quickly with $Z$ for redder colors. Also, histograms for redder colors truncate at smaller values of $Z$. Notably, the stellar density is roughly the same for all curves near $Z=0$, so the differences between the curves is a selection effect. Red stars tend to be small, low-luminosity stars, and so can only be detected close to Earth due to the SDSS apparent magnitude limit. Bluer stars, however, tend to be higher luminosity and therefore can be detected much farther away.

I would expect larger systematic error in the reddest bin because most of those stars are very faint and therefore have large measurement errors due to Poisson noise. Also there could potentially be contamination from evolved, giant stars in the reddest color bin.

## Problem 3: Compare $\ln \rho$ vs. $Z$ for blue stars with low vs. high metallicity

```python
#Problem 3

fig, ax = plt.subplots(figsize=(10,8))
ax.set_xlabel(r'$Z$')
ax.set_ylabel(r'$\ln \rho$')

bin_edges, ln_rho = get_ln_rho(Z, 0.2, 0.4, bins=30, low_metallicity=-1.0)
ax.plot(bin_edges[1:], ln_rho, linestyle='steps', label=r'$[Fe/H] > -1.0$')

bin_edges, ln_rho = get_ln_rho(Z, 0.2, 0.4, bins=30, high_metallicity=-1.0)
ax.plot(bin_edges[1:], ln_rho, linestyle='steps', label=r'$[Fe/H] < -1.0$')

plt.legend()
```

This plot shows that low metallicity stars lie at smaller distances and their number density drops off more quickly with $Z$ than that of the high metallicity stars. Low metallicity stars typically formed earlier (when stellar evolution had had less time to pollute the ISM with metals), and therefore tend to be older, redder, and fainter (note though that these stars are all within a relatively narrow color range, so the low metallicty stars can't be TOO much older). Because they are fainter, they are not detected at large distances, hence the truncation of the low metallicity star histogram at $Z \sim 5 \times 10^4$ pc. Higher metallicity stars formed more recently on average, therefore tend to be brighter and detectable at larger distances.

## Problem 4: Number of stars in bins of $r$ magnitude for low vs. high metallicity

```python
#Problem 4 

high_met_samp = data[:,col=='r'][:,0][(gr > 0.2) * (gr < 0.4) * (metallicity > -1.0)]
low_met_samp = data[:,col=='r'][:,0][(gr > 0.2) * (gr < 0.4) * (metallicity < -1.0)]

fig, ax = plt.subplots(figsize=(10,8))

hist, bin_edges = np.histogram(high_met_samp, bins=70)
ax.plot(bin_edges[1:], hist, linestyle='steps', label=r'$[Fe/H] > -1.0$')

hist, bin_edges = np.histogram(low_met_samp, bins=70)
ax.plot(bin_edges[1:], hist, linestyle='steps', label=r'$[Fe/H] < -1.0$')

ax.set_xlabel(r'$r$')
ax.set_ylabel(r'$N$')
plt.legend(loc=2)
```

This plot is consistent with my previous conclusion that low metallicity stars are typically fainter than high metallicity stars. The bulk of stars with low metallicity lie near the faint end of the survey, whereas the high metallicty stars are relatively evenly distributed across the full range of $r$ magnitude.

The low metallicity curve rises roughly exponentially up to roughly $r \sim 20$, where it drops off rapidly. The high metallicity curve looks like a mixture of two Gaussians, since the distribution is clearly bimodal.

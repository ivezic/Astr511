Compute absolute magnitudes of SDSS stars using eqs. A2, A3, A7 from Ivezic et al. (2008):

$\Delta M_r(\mathrm{[Fe/H]}) = 4.50 - 1.11 \mathrm{[Fe/H]} - 0.18 \mathrm{[Fe/H]}^2 \\
M_r (g-i, \mathrm{[Fe/H]}) = M_r^0(g-i) + \Delta M_r(\mathrm{[Fe/H]}) \\
M_r^0 (g-i) = -5.06 + 14.32(g-i) - 12.97(g-i)^2 + 6.127(g-i)^3 - 1.267(g-i)^4 + 0.0967(g-i)^5.$

To obtain [Fe/H], use the relation from Bond et al. (2010):

$\mathrm{[Fe/H]} = -0.6$ for $(g-r)>0.6$ (redder stars), and 

$\mathrm{[Fe/H]} = A + Bx + Cy + Dxy + Ex^2 + Fy^2 + Gx^2y + Hxy^2 + Ix^3 + Jy^3$ for $(g-r)<0.6$, 
where $x=(u-g)$ and $y=(g-r)$ here; the coefficients (A-J) = (−13.13,14.09,28.04,−5.51,−5.90,−58.68, 9.14,−20.61, 0.0, 58.20).

```python
ug = sdss[:, 4] - sdss[:, 5]
gr = sdss[:, 5] - sdss[:, 6]
```

```python
#calculate absolute r magnitude
def computeMr(retfeh=False):
    Mr0 = -5.06 + 14.32*(sdss[:, 5]-sdss[:, 7]) - 12.97*(sdss[:, 5]-sdss[:, 7])**2 \
        + 6.127*(sdss[:, 5]-sdss[:, 7])**3 - 1.267*(sdss[:, 5]-sdss[:, 7])**4 \
        + 0.0967*(sdss[:, 5]-sdss[:, 7])**5

    coeff = np.array([-13.13,14.09,28.04,-5.51,-5.90,-58.68, 9.14,-20.61, 0.0, 58.20])
    feh = coeff[0] + coeff[1]*ug + coeff[2]*gr + coeff[3]*ug*gr + coeff[4]*ug**2 + coeff[5]*gr**2 \
        + coeff[6]*ug**2*gr  + coeff[7]*ug*gr**2  + coeff[8]*ug**3 + coeff[9]*gr**3
    red = (gr>0.6)
    feh[red] = -0.6
    Mr = Mr0 + 4.50 - 1.11*feh - 0.18*feh**2
    if retfeh:
        return feh, Mr
    return Mr

#compute the distance via distance modulus
def distmod(obsmag, absmag):
    Z = 10**((obsmag - absmag)/5.) * 10 # in pc
    return Z
```

```python
#approximate distance from galactic plane = distance from us for b>80deg
feh, Mr = computeMr(retfeh=True)
z = distmod(sdss[:, 6], Mr)
```

```python
# find optimal binwidth from Freedman-Diaconis Rule 
def fdrule(x):
    binwidth = 2.7*np.std(x) / len(x)**(1./3.)
    return binwidth

# find the volume of spherical slice, where 
# dV = r^2 sin(theta) d(phi) d(theta) dr 
def getVol(theta, ri, rf):
    vol = 2.*np.pi*(np.cos(0.)-np.cos(theta*np.pi/180))*(rf**3-ri**3)/3.
    return vol
```

```python
# compute the stellar density given a set of color, metallicity cuts
def getRho(hi, lo, z, nbins=None, retedge=False, feh=None, fehcut=None):
    if nbins is None:
        binwidth=fdrule(z)
        nbins = int(np.ceil((z.max()-z.min())/binwidth))
    #print hi, lo, nbins
    cut = (gr<hi) * (gr>lo)
    if fehcut == 'low':
        cut = cut * (feh<-1.0)
    elif fehcut == 'high':
        cut = cut * (feh>-1.0)
    zcount, zedge = np.histogram(z[cut], bins=nbins)
    if retedge:
        return zedge
    vol = np.zeros(len(zcount))
    vol = getVol(10., zedge[:-1], zedge[1:])
    rho = zcount/vol
    return (zedge[:-1]+zedge[1:])/2., rho
```

### 1. ### 
For stars with $0.2< g−r <0.4$, plot ln(ρ) vs. Z, where ρ is the stellar number density in a given bin (e.g. look at Figs. 5 and 15 in Juric et al. 2008, ApJ, 673, 864 for similar examples). You can approximate ρ(Z) = N(Z)/V(Z), where N(Z) is the number of stars in a given bin, and V(Z) is the bin volume (note that the solid angle is ∆Ω∼314 deg◦). What is the Z range where you believe the results, and why?

The stellar log-density distribution drops steeply for the first $\sim$5 kpc, and then more gradually until $\sim$40 kpc, where the slope flattens slightly. Beyond 60 kpc there is little data. I was not sure what was meant by "believing the results," so I interpretted it two ways. 1) Trusting that the observed stellar density is tracing the true stellar density/distribution of all stars fiducially. 2) Trusting that the observed stellar density is capturing some essence of the overall stellar density (i.e., two intrinsically different stellar distribuions - exponential vs. power-law). 

To address the first point, it became more clear while doing the rest of the problem set that the observations more or less reflect the true distribution of stars for $Z\lesssim 5-6$ kpc. Namely, slicing the data in color demonstrated that within 5 to 6 kpc, there are more redder (later-type) than bluer MS stars nearby, as expected. 

To address the second point, the density profile does show remarkable structure. Within 5-6 kpc, the density decrease is sharp (exponential, indicative of disk). Beyond this, the density decrease is more gradual (power-law, indicative of halo). However, at around 40 kpc, the density decrease seem to flatten off from 5-40 kpc range. It's unclear why this should physically occur, so in this sense, I would not trust the density profile beyond 40 kpc.

```python
nbins=100
fig = plt.figure(figsize=(8, 6))
z1, rho1 = getRho(0.4, 0.2, z, nbins=nbins)
ax = fig.add_subplot(111)
inset_ax = fig.add_axes([0.5, 0.5, 0.35, 0.35])
ax.plot(z1/1000., np.log(rho1), ls='steps')
ax.set_ylim((-29, -9))
ax.plot((5, 5), ax.get_ylim(), 'k:', label='sharp drop off')
ax.plot((40, 40), ax.get_ylim(), 'r:', label='gradual decline')
inset_ax.plot(z1/1000., np.log(rho1), ls='steps')
inset_ax.set_xlim((0,20))
inset_ax.set_ylim((-17, -9))
inset_ax.plot((5,5),inset_ax.get_ylim(),'k:')
ax.set_xlabel('$Z\ \\mathrm{(kpc)}$')
ax.set_ylabel('$\ln \\rho(Z)$')
ax.set_title('$\ln \\rho(Z)\ \mathrm{vs.}\ Z.\ \mathrm{for}\ 0.2<g-r<0.4$', fontsize=20)
ax.legend(loc=8)
```

### 2. ###
Add ln(ρ) vs. Z for stars with $0.4< g−r <0.6$, $0.6< g−r <0.8$, and $0.8< g−r <1.0$ (you can rescale all curves to the same value at some fiducial Z, or leave them as they are). Discuss the differences compared to the $0.2< g−r <0.4$ subsample. Why do we expect larger systematic errors for $0.8< g−r <1.0$ than for the adjacent bin with $0.4< g−r <0.6$?

At short distances (Z $\lesssim 5-6$ kpc, by eye), the curves corresponding to redder colors actually exceed the $0.2<g-r<0.4$ stellar density curve (see inset). This makes sense since later-type stars are more abundant and so within a given volume we should observe more red stars than blue (given that they are above the detection limit). The different stellar log-density profiles show similar decreasing trends for these low $Z$ values. Beyond this height $Z_\mathrm{crit}$, stellar density profiles corresponding to redder colors drop off more quickly. In other words, $\rho$ for the bluest stars in the sample is systematically higher beyond $Z=6$ kpc. 

This apparent over-abundance of bluer stars relative to redder stars per unit volume at larger galactic heights is not physical. Rather, it is due to the fact that bluer MS stars are more intrinsically bright, and so more readily detected. Redder MS stars, on the other hand, are intrinsically fainter. Therefore they will not be detected if they are far enough that they appear dimmer than the faintness detection limit. I'm not sure what the red giant contamination fraction is...

We expect larger systematic errors for $0.8 < g−r <1.0$ than for the adjacent bin with $0.4< g−r <0.6$. The former comprise of redder, fainter stars than the latter, and fewer photons mean larger Poisson noise. More importantly for systematic errors, the metallicity relation only holds for stars with $g-r<0.6$. For stars redder than this, we assumed a floor on the [Fe/H] value, and thus only have a limit for $M_r$ and $Z$. 

```python
hi = [0.4, 0.6, 0.8, 1.0]
lo = [0.2, 0.4, 0.6, 0.8]
nbins=100
label = ['$'+str(lo[ii])+'<g-r<'+str(hi[ii])+'$' for ii in range(len(hi))]
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)
inset_ax = fig.add_axes([0.6, 0.3, 0.25, 0.25])
for ii in range(len(hi)):
    if ii==0:
        binedges=getRho(hi[ii], lo[ii], z, nbins=nbins, retedge=True)
    z1, rho1 = getRho(hi[ii], lo[ii], z, nbins=binedges)
    ax.plot(z1/1000., np.log(rho1), ls='steps', label=label[ii])
    inset_ax.plot(z1/1000., np.log(rho1), ls='steps', label=label[ii])
ax.set_xlabel('$Z\ \\mathrm{(kpc)}$')
ax.set_ylabel('$\ln \\rho(z)$')
ax.plot([5, 5], plt.gca().get_ylim(), 'k:', label='$Z_{\\mathrm{crit}}$')
ax.set_ylim((-28, -8))
inset_ax.set_ylim((-15, -8))
inset_ax.set_xlim((0, 7))
ax.legend(fontsize=18)
```

### 3. ###
For subsample with $0.2< g−r <0.4$, separate stars into low-metallicity sample, [Fe/H]$<−1.0$, and high-metallicity sample, [Fe/H]$>−1.0$. Compare their ln(ρ) vs.Z curves. What do you conclude?

The high metallicity stellar density curve decreases sharply until about $Z\sim5-6$ kpc, where it then decreases more gradually. The low metallicity curve, on the other hand, steadily decreases with distance. At about 20 kpc, there is a cross-over and high metallicity stars appear to have a higher density. This is again, an apparent effect and not physical (in fact we expect the halo to be dominated by older, redder stars). Observations are missing the low metallicity stars because they belong to the oldest population, which means they are low mass and very faint, beyond the detection limit at such distances. 

Consequently, again, we see that the observed stellar density traces the true stellar density only at small distances - very close to the galactic plane exist high and low metallicity stars, above the galactic plane there are more older (redder, low metallicity) stars.

```python
fehcut=['high', 'low']
plt.figure(figsize=(10, 8))
for ii in range(len(fehcut)):
    z1, rho1 = getRho(hi[0], lo[0], z, nbins=binedges, feh=feh, fehcut=fehcut[ii])
    plt.plot(z1/1000., np.log(rho1), ls='steps', label=fehcut[ii]+'-metallicity')
plt.xlabel('$Z\ \\mathrm{(kpc)}$')
plt.ylabel('$\ln \\rho(z)$')
plt.title('Stellar density for $0.2<g-r<0.4$', fontsize=18)
plt.plot((5,5), plt.gca().get_ylim(), 'k:', label='$Z_{\\mathrm{crit}}$')
plt.legend()
```

### 4. ###
For these low-metallicity and high-metallicity samples, plot and compare their differential $r$ band magnitude distributions (i.e. the number of sources per unit magnitude, in small, say 0.1 mag wide, $r$ bins). What do you conclude? How would you numerically describe these curves (i.e. what kind of functional form for the fitting functions would you choose)?

The low metallicity sample density increases almost exponentially with increasing magnitude values (decreasing brightness) until $r\sim20$ mag, where it turns over and decreases at the faintest magnitudes. The high metallicity sample is comparatively more constant, with a generally increasing trend with dimmer $r$ brightness (except for an apparent trough at intermediate $r\sim18-19$ values. The dearth of low metallicity stars at bright $r$ mags relative to faint $r$ end is consistent with what we know about stellar populations in the Milky Way: low metallicity (first generation, old) stars are more common in the halo. 

I would use an exponential fit to the low metallicity (until $r\sim20$ mag). The high metallicity curve looks almost bi-modal, so I would fit it with a double Gaussian. 

```python
def getdiffR(feh, r, binsize, fehcut=None):
    cut = (gr<0.4) * (gr>0.2)
    nbins = int(np.ceil((r[cut].max()-r[cut].min())/binsize))
    nbins = r[cut].min() + np.arange(nbins)*binsize
    if fehcut == 'high':
        cut = cut * (feh>-1.0)
    elif fehcut == 'low':
        cut = cut * (feh<-1.0)
    rcount, redge = np.histogram(r[cut], bins=nbins)
    return (redge[:-1]+redge[1:])/2., rcount
```

```python
plt.figure(figsize=(10,8))
for ii in range(len(fehcut)):
    r1, rcount1 = getdiffR(feh, sdss[:, 6], 0.1, fehcut=fehcut[ii])
    plt.plot(r1, rcount1, ls='steps', label=fehcut[ii]+'-metallicity')
plt.xlabel('$r\ \\mathrm{(mag)}$')
plt.ylabel('$\\mathrm{Distribution}$')
plt.legend(loc=2)
```

### 5. ###
What should be the faint $r$ band limit for a survey to be able to map the ln(ρ) vs. Z profile out to 100 kpc using main-sequence stars? Assume the same color distribution as for the SDSS sample. For a solid angle of 1 deg$^2$, how many stars with $0.2< g−r <0.4$ would you expect with distances between 90 kpc and 100kpc? Assume whatever additional information you need to solve this problem (not all required information is provided here).

Below I plotted $\log_{10}(Z)$ in kpc as a function of $r$ mag. I applied a least-squares fit to the data to extrapolate the expected $r$ at 100 kpc, which predicts that MS stars should have $r \sim 24.7$ mag. Since there is significant scatter about the best-fit line, I computed the standard deviation between the data and best-fit line, and extrapolated the corresponding $r$ value at +/- 3$\sigma$. These are effectively the conservative and optimistic limits, at 25.3 and 24.2 mag. So to be on the safe side, the $r$ band limit should be 25-25.5 mag.

To compute the expected number of stars with $0.2< g−r <0.4$ between 90-100 kpc, I calculated $\ln N(Z)$ as a function of $Z$, similar to question 1. Then I fit a least-squares line to $\ln N$ for $6<Z<40$ kpc, the range in which the stellar density drops off uniformly, as was also established in question 1. Finally I calculated the area under the best-fit line for Z = 90-100 kpc, and extrapolated that we would expect to find $<1$ MS star. This is of course the value with the detection bias folded into it. To find the true number of stars, one would have to account for the detection deficiency as a function of distance. 

```python
colorcut = (gr<0.4)*(gr>0.2) #* (Z<70000)
plt.figure(figsize=(8,6))
plt.plot(sdss[:, 6][colorcut], np.log10(Z[colorcut]/1000.), 'r.', label='data')
x = sdss[:, 6][colorcut]
y = np.log10(Z[colorcut]/1000.)
A = np.vstack([x, np.ones(len(x))]).T
m, b = np.linalg.lstsq(A, y)[0]
xnew = np.linspace(13, 27, 100)
ynew = m*xnew+b
rlim_ind = np.where(abs(ynew-2)==abs(ynew-2).min())
scatter = np.std(y-(m*x+b))
plt.plot(xnew, ynew, 'k--', label='Least-squares fit')
plt.fill_between(xnew, ynew-3*scatter, ynew+3*scatter, color='k', alpha=0.3)
plt.text(22, 2.3, '$3\sigma$ scatter to fit', rotation=25)
plt.plot(xnew[rlim_ind], ynew[rlim_ind], 'bo', ms=8, label='faint $r$ limit~'+str(xnew[rlim_ind][0])[:4])
plt.xlim(13, 27)
plt.xlabel('$r\ \\mathrm{(mag)}$')
plt.ylabel('$\log(Z\ \\mathrm{[kpc]})$')
plt.legend(loc=2)
print "faint r, upper limit, lower limit = ", xnew[rlim_ind][0], \
    xnew[np.where(abs(ynew+scatter-2)==abs(ynew+scatter-2).min())[0]][0], \
    xnew[np.where(abs(ynew-scatter-2)==abs(ynew-scatter-2).min())[0]][0]
```

```python
nbins=100
fig = plt.figure(figsize=(8, 6))
z1, rho1 = getRho(0.4, 0.2, z, nbins=nbins)
zcount, zedge = np.histogram(z[(gr<0.4)*(gr>0.2)], bins=nbins)
z1*=1e-3
cut = (z1<40) * (z1>6)
x = z1[cut]
y = np.log(zcount)[cut]
A = np.vstack([x, np.ones(len(x))]).T
m, b = np.linalg.lstsq(A, y)[0]
xnew = np.linspace(5, 100, 10000)
ynew = m*xnew+b
ax = fig.add_subplot(111)
ax.plot(z1, np.log(zcount), ls='steps')
ax.plot(xnew, ynew, 'r--')

#now extrapolate ln(N) at Z=90-100 kpc
N95= np.trapz((np.exp(ynew[xnew>90])), xnew[xnew>90])
print "Expected number of stars between 90-100 kpc with $0.2<g-r<0.4$ = ", np.round(N95, decimals=3)
ax.plot((5, 5), ax.get_ylim(), 'k:', label='sharp drop off')
ax.plot((40, 40), ax.get_ylim(), 'r:', label='gradual decline')
ax.fill_between((90,100),(-5.3, -7), (-8,-8), color='g', alpha=0.3)
ax.text(75, -7, '$N_*=$'+str(np.round(N95, decimals=3)), fontsize=18)
ax.set_xlabel('$Z\ \\mathrm{(kpc)}$')
ax.set_ylabel('$\ln\ N(Z)$')
ax.set_title('$\ln\ N(Z)\ \mathrm{vs.}\ Z.\ \mathrm{for}\ 0.2<g-r<0.4$', fontsize=20)
ax.legend(loc=1)
```

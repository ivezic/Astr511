### Loading  the SDSS photometry for 600 000 stars  

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.size'] = 17

File = 'Astr511HW1data.dat'
# Save the column names
colnames = open(File, 'r').read().splitlines()[0][1:].split()

# Load the rest of the file 
datatable = np.genfromtxt(File)

# Form a structure to hold the data 
data = {}
for label, column in zip(colnames, datatable.T):
    data[label] = column   # using  http://stackoverflow.com/questions/209840/map-two-lists-into-a-dictionary-in-python 
```

### Equations to calculate absolute magnitude :

$$[\text{Fe/H}] = A + Bx + Cy + Dxy + Ex^2 + Fy^2 + Gx^2y + Hxy^2 + Ix^3 + Jy^3 \tag{1}$$

with $x = (u-g)$ and $y = (g-r)$

$$ \Delta M_r([\text{Fe/H}]) = 4.50 - 1.11[\text{Fe/H}] - 0.18[\text{Fe/H}]^2 \tag{A2}$$

$$ M_r(g-i, [\text{Fe/H}]) = M_r^0(g-i) + \Delta M_r([\text{Fe/H}]) \tag{A3}$$

$$ M_r^0(g-i) = -5.06 + 14.32x - 12.97x^2 + 6.127x^3 - 1.267x^4 + 0.0967x^5 \tag{A7}$$

with $x = (g-i)$

From equations above, to calculate absolute magnitude I first compute metallicity using an empirical fit to the $u-g$ and $g-r$ SDSS colors. However, this fit is only valid for   $g-r < -0.6$, so I first need to make a mask for all rows which fulfill the criterion. Wherever $g-r > -0.6$, $[Fe/H] = -0.6$, following  [Bond et al. 2010 (ApJ, 716, 1)](http://adsabs.harvard.edu/abs/2010ApJ...716....1B).  Given  $[Fe/H]$, I calculate $\Delta M_{r}$ (eq. A2), and then $M_{r}^{0}$ (eq. A7). I combine them to get $M_{r}$ via eq. A3. 

### Find the absolute magnitude 

```python
# Define x and y for eq. (1)
x = data['u'] - data['g']
y = data['g'] - data['r']  # coefficients from Bond et al. 2010
A, B, C, D, E, F, G, H, I, J = [-13.13, 14.09, 28.04, -5.51, -5.90, -58.68, 9.14, -20.61, 0.0, 58.20]  

###
###   METALLICITY  
###

# Make the mask 
useEquation = y < 0.6
# Make a new key in the dictionary for metallicity 
data['FeH'] = np.zeros_like(x,dtype=float)

# only where it is  valid fill in the rows of 'FeH' column with values from eq. (1)
data['FeH'][useEquation] = (A + B*x + C*y + D*x*y + E*x**2 + F*y**2 + G*y*x**2 + 
                                 H*x*y**2 + I*x**3 + J*y**3)[useEquation]

# At those places where the condition is not fulfilled, 
# the metallicity is constant (for redder Stars )
data['FeH'][np.invert(useEquation)] = -0.6

###
###  ABSOLUTE MAGNITUDE  
###

# Use eq. (A2) for delta_M_r
data['del_M_r'] = 4.50 - 1.11*data['FeH'] - 0.18*data['FeH']**2

# Use eq. (A7) to get M_r^0 . Note that Now x means something different than in eq. (1) !  
x = data['g'] - data['i']
data['M_r0'] = -5.06 + 14.32*x - 12.97*x**2 + 6.127*x**3 - 1.267*x**4 + 0.0967*x**5

# Use eq. (A3) to find Absolute magnitude  M_r
data['M_r'] = data['M_r0'] + data['del_M_r']
```

### Calculate the distance to stars

Given that $b>80^\circ$, the distance from the galactic plane : $Z$, and the distance from us, $D$, are approximately the same.   

```python

# Assuming D~Z, find the distance in parsecs
data['D'] = 10 * 10**((data['r'] - data['M_r'])/5.)
```

## Part 1.

For stars with a given  range of  color   $0.2 < g-r < 0.4$ I plot  the star distance  vs. natural log of the stellar number density in a given bin  ( $Z$ vs. $\ln\rho$ , as in Fig. 15 reproduced below). This is similar to the work done in [Juric et al. 2008, ApJ, 673, 864](http://iopscience.iop.org/0004-637X/673/2/864/), Figs. 5 and 15 are reproduced below. I approximate  $\rho(Z) \approx N(Z) / V(Z)$, where $N(Z)$ is the number of stars in a given bin, and $V(Z)$ is the bin volume.  <img src="medium-fg15_online.gif">

To use the formula given for $\rho(Z)$ I need to find the $V(Z)$, volume at a given distance. Since we are taking a cone survey, I can approximate the volume as a wedge  $$ V(Z) \approx V(D) = \int_0^{2\pi} d\phi \int_0^{\theta '} d\theta \int_{D_0}^{D_1} dD \, (D^2 )\sin\theta
= \frac{2\pi}{3} \left[ D^3 \right]_{D_0}^{D_1} \left[ \cos\theta \right]_{\theta'}^{0} $$

### Plotting the volume density 

```python
# Limit my rows to only those stars with  0.2 < g-r < 0.4
g_r = data['g'] - data['r']
color_mask = (0.2 < g_r)*(g_r < 0.4)

# Calculate the histogram in Distance calculated from distance modulus
nbins = 60
N_z, edges = np.histogram(data['D'][color_mask], nbins, range=[0,7e4])

# Calculate the center of each bin : take an average 
center  =  [ 0.5 * edges[i] + edges[i+1]  for i in range(len(edges)-1) ]  # in parsecs!  

# Convert to kiloparsecs
center_kpc = 0.001 * np.array(center)

# Calculate the V(Z) using the equation above 
V_z = [2.0*np.pi/3.0 * (edges[i+1]**3 - edges[i]**3) * (np.cos(0) - np.cos(np.radians(10))) for i in range(len(edges) - 1)] 
rho_z = N_z / V_z

# Plot the  Z vs ln (rho) : 
plt.clf()
plt.plot(center_kpc, np.log(rho_z), ls = 'steps', lw=2)
plt.ylabel(r'$\ln \rho(Z) \, \approx \, \ln \left( N(Z)/V(Z) \right)$')
plt.xlabel('$Z$ [kpc]')
plt.title('Stellar number density')
```

The range over which I believe in the results depends on various errors, but if these were not known, the estimate is based on number of stars per bin, and if that number becomes very small, then the estimate may be unreliable.  Thus I plot below number of stars per bin:

```python
plt.clf()
plt.plot(center_kpc, N_z, ls = 'steps', lw=2)
plt.ylabel(' N(Z)')
plt.xlabel('$Z$ [kpc]')
plt.title('Number of stars per bin')
plt.axhline(0.1 * np.max(N_z), color = 'red', lw=2)
```

Above I marked the $10\%$ line, beyond which the number of stars per bin is less than $10\%$ of the maximum value. With small number statistics there may be problems, so my estimate is that beyond about $30$ kpc the estimate may get progressively unreliable. 

## Part 2.

```python
# This time define a function, since I need to plot the same thing three times 

def nd_plotting(g=data['g'], r = data['r'], gr_min=0.2 , gr_max=0.4, nbins=60, C='blue'  ):
    '''
    Plotting the distance vs  stellar number density , 
    for  a given color range of SDSS stars,    lower_limit < g-r < upper_limit . 
    The filters used are  g and  r ,  and  the default  
    (g-r)_lower_limit = 0.2 ,   (g-r)_upper_limit = 0.4
    '''
    
    # Limit my rows to only those stars with  0.2 < g-r < 0.4
    g_r = data['g'] - data['r']
    color_mask = (gr_min < g_r)*(g_r < gr_max)

    # Calculate the histogram in Distance calculated from distance modulus
    N_z, edges = np.histogram(data['D'][color_mask], nbins, range=[0,7e4])

    # Calculate the center of each bin : take an average 
    center  =  [ 0.5 * edges[i] + edges[i+1]  for i in range(len(edges)-1) ]  # in parsecs!  

    # Convert to kiloparsecs
    center_kpc = 0.001 * np.array(center)

    # Calculate the V(Z) using the equation above 
    V_z = [2.0*np.pi/3.0 * (edges[i+1]**3 - edges[i]**3) * (np.cos(0) - np.cos(np.radians(10))) for i in range(len(edges) - 1)] 
    rho_z = N_z / V_z

    # Plot the  Z vs ln (rho) : 
    plab = str(gr_min)+'< g-r <'+str(gr_max)
    plt.plot(center_kpc, np.log(rho_z), ls = 'steps', lw=2, label = plab , color = C)



# Clean the canvas 
plt.clf()
fig1 = plt.figure(figsize=(12,10))
# First plot the old plot : 0.2 < g-r < 0.4 
nd_plotting( gr_min=0.2 , gr_max=0.4 )

# Second overplot the 0.4 < g-r < 0.6
nd_plotting( gr_min=0.4 , gr_max=0.6 , C='green')

# Third overplot: 0.6 < g-r < 0.8
nd_plotting( gr_min=0.6 , gr_max=0.8, C='orange' )

# Fourth overplot :  0.8 < g-r < 1.0
nd_plotting( gr_min=0.8 , gr_max=1.0 , C='red')

# Add labels and title  
plt.ylabel(r'$\ln \rho(Z) \, \approx \, \ln \left( N(Z)/V(Z) \right)$')
plt.xlabel('$Z$ [kpc]')
plt.legend()
plt.title('Stellar number density')
plt.show()
```

On the plot above we see that the stars with highest values of  $g-r$, i.e. much brighter in $r$ than in $g$ - the reddest stars, have also a population that falls off quickest with galactocentric distance. This may be due to a selection effect- namely, I did not correct for the fact that redder stars are also on average dimmer than the young, blue stars. Thus the blue main sequence stars would be visible over large distances (following http://zebu.uoregon.edu/~soper/Stars/trl.html and http://astro.physics.uiowa.edu/~kaaret/s09/L12_starsmainseq.pdf ). 

Errors for redder stars would  be bigger because we receive less flux, for the same amount of intstrumental uncertainty, background noise, etc. Thus Signal-to-Noise for blue stars should be bigger than for redder stars, as stated.

I conclude that for bluer stars, high metallicity ones tend to be scattered more at higher galactocentric distances. I can also see that the metal-poor (red on the image)  population  has a cusp in the center, and falls off very rapidly, and then has a break, whereas the metal-rich (blue on the image) population follows a more uniform power-law.  Perhaps this means that the metal-rich population comes from a starburst that recycled the material that came from an edge-on collision, which left at lower radii the lower-metallicity ones, that did not take part in the collision.

## Part 5.

To find out how deep we would need to go with our survey to map  the  $\ln \rho$ vs. $Z$ profile, I first find how the r-band data  behaves with distance $Z$, by plotting  $Z$ vs $r$-band within the color range $0.2<g-r<0.4$, which falls off least steeply. 

### Faint r-band limit 

```python
# make colour mask  on g-r 
g_r = data['g'] - data['r']
color_mask = (0.2 < g_r)*(g_r < 0.4)

# apply the mask to r-band 
rband = data['r'][color_mask]

# calculate the distance , answer in kpc  
logZ = np.log10(0.001 * data['D'][color_mask])   

# Plot the r-band and the desired depth 
fig1, ax1 = plt.subplots(figsize=(12,10))  
ax1.scatter(logZ, rband, s=0.2, alpha=0.2)
ax1.set_xlabel(r'$log(Z)[kpc]$')
ax1.set_ylabel('r-band  magnitude')
ax1.set_title('R-band stars')
ax1.annotate('Desired limit: \n100 kpc',xy=(1.3,17), color='blue')
ax1.axvline(x=2.0,ls='--',lw=2)



# Overplot the linear fit  using http://docs.scipy.org/doc/numpy/reference/generated/numpy.polyfit.html
# y = a*x + b

a,b = np.polyfit(logZ, rband, 1)

# at x=2 : desired faint limit of distance , find the value of y : 
y_faint =a*2.0 + b
ax1.axhline(y=y_faint, ls='--', lw=2 )
text = 'Faint limit : r='+str(y_faint)[:5]
ax1.annotate(text, xy=(0.5,y_faint+.5), color='blue')

# Plot the linear fit 
x_args = np.linspace(0, 2, 10)
y_vals = a * x_args + b
ax1.plot(x_args, y_vals, 'r', lw=2)
plt.show()
```

Thus by extrapolating the existing data,  I find that to reach $100$ kpc, the survey should be able to detect stars as faint as  $24.07$ magnitude. 

### Number of stars between $90$ and $100$ kpc

To find how many stars we should expect between  $90$ and $100$ kpc, I use similar method as above, but I extrapolate data from Part 1., with  $0.2 < g-r < 0.4$. 

```python
# Same code as in part 1 : plotting the volume density 
# Limit my rows to only those stars with  0.2 < g-r < 0.4
g_r = data['g'] - data['r']
color_mask = (0.2 < g_r)*(g_r < 0.4)

# Calculate the histogram in Distance calculated from distance modulus
nbins = 60
N_z, edges = np.histogram(data['D'][color_mask], nbins, range=[0,7e4])


# Calculate the center of each bin : take an average 
center  =  [ 0.5 * edges[i] + edges[i+1]  for i in range(len(edges)-1) ]  # in parsecs!  

# Convert to kiloparsecs
center_kpc = 0.001 * np.array(center)

# Plot the  Z vs ln (N(z)) : 
fig3, ax3 = plt.subplots(1, figsize=(12,10))  
ax3.plot(center_kpc, np.log(N_z), ls = 'steps', lw=2, color='green')
ax3.set_ylabel(r'$\ln N(Z)$')
ax3.set_xlabel('$Z$ [kpc]')
ax3.set_title('Stellar number density')


# Fit a straight line only to those stars up to about 30 kpc (what I determined to be the reliable range in Part.1 )
# y = ax + b 
distance_mask = (center_kpc > 10 )*( center_kpc < 30)
a,b = np.polyfit(center_kpc[distance_mask], np.log(N_z[distance_mask]), 1)

# Plot the linear fit 
x_args = np.linspace(10, 101, 10)
y_vals = a * x_args + b
ax3.plot(x_args, y_vals, 'purple', lw=2)
plt.show()

# Integrate under the curve between 90 and 100 kpc 
# I took the natural log which is why I need to take exp(y_vals) to get star counts
star_number = np.trapz(np.exp(y_vals[x_args > 90]), x_args[x_args > 90])
```

```python
import numpy as np
import pandas as pd
import prettyplot as prt
import matplotlib.pyplot as plt
from astroML import plotting, stats
% matplotlib inline
```

## Load files and merge in pandas ##

```python
dfObs = pd.read_csv('LSSTsimWDobs60.dat', delim_whitespace=True, header=0, index_col=False)
dfTrue = pd.read_csv('LSSTsimWDtruth60.dat', delim_whitespace=True, header=0, index_col=False)
# Merge observation and "truth" into one data frame
df = pd.merge(dfObs, dfTrue, left_index=True, right_index=True)
```

# Part A #

## Gold Parallax Sample ##

```python
dfGold = df[df.piObs/df.piErr > 10.] # pi SNR > 10
```

```python
modObs = 5.*np.log10(100./dfGold.piObs) # observed distance modulus
modTrue = dfGold.r-dfGold.Mr # true distance modulus
diff = modObs - modTrue # difference
```

```python
median, sigma_g = stats.median_sigmaG(diff)
mean, sigma = stats.mean_sigma(diff)
clip = diff[np.abs(diff-median) < 5.*sigma_g] # clip at 5 sigma
```

## Statistics ##

```python
print 'Median: %.4f' % median
print 'Mean: %.4f'   % mean
print 'Sigma_G %.3f' % sigma_g
print 'Sigma: %.3f'  % sigma
```

## Histogram ##

```python
#plt.style.use('bmh')
fig, ax = plt.subplots(figsize=(10,8))
prt.font(size=16)

#prt.font(size=16)
plotting.hist(clip,bins='knuth')
#prt.prettyfy(ax, xmaj=0.2, xmin=0.05, ymaj=200, ymin=25, xfmt='%.1f',yfmt='%d')
ax.set_xlim(-0.75, .65);ax.set_ylim(0, 1100)
ax.set_xlabel('$\Delta$ Dist. Mod. (Observed - Truth)')
ax.set_ylabel('Number of Stars')
plt.show()
```

I've clipped the distribution at 5$\sigma$ and plotted the histogram using Knuth's rule to determine the bin size. The scatter is about what I would expect. For most of the stars the observed luminosity is differing from the true luminosity by about 0.15 magnitues (~15%). All of the stars in the sample have their distances measured to within 10%, but luminosity depends on distance squared, so I assume the additional scatter is due to the error propagation.

The distribution differs from a Gaussian, as $\sigma_G$ does not equal $\sigma$ assuming a Gaussian. This is not surprising, becuase even if the parallax errors are normally distributed, the distance errors will not be due to the $1/\pi$ dependence.

The cap with b < 60 has a solid angle of .842 steradians, or 2,760 square degrees. So there will be about 7 times more white dwarfs in the full survey area, about 5.5 million.

# Part B #

## Hydrogen WD Sample ##

```python
hyd = dfGold[dfGold['T']==1] # select hydrogen WDs
gr = hyd.g - hyd.r
Mobs = hyd.rObs - 5.*np.log10(100./hyd.piObs)
```

```python
# Calculate the median and standard deviation of the binned data
medgr, bins = stats.binned_statistic(gr, Mobs, bins=np.arange(-0.4, 1.0, 0.1), statistic='median')
stdgr, bins = stats.binned_statistic(gr, Mobs, bins=np.arange(-0.4, 1.0, 0.1), statistic=np.std)
x = bins[:-1]+0.05
```

```python
from astroML.linear_model import PolynomialRegression

degrees = range(1,6)
AIC = []; BIC = []
for deg in degrees:
    
    polynomial = PolynomialRegression(deg)
    polynomial.fit(x[:,None], medgr)
    fit = polynomial.predict(x[:,None])
    
    k = deg                      # number of parameters = degree of polynomial
    N = medgr.shape[0]           # number of data points
    SSR = np.sum((medgr - fit)**2.) # sum of square residuals
    sig2 = SSR / N  # variance (sigma squared)
    lnL = np.log((1.0/np.sqrt(2*np.pi*sig2))** N * np.exp(-SSR/(sig2*2.0) )) # log likelihood
    
    aic = 2*k - 2*lnL
    AIC.append(aic)
    bic = k*np.log(N) - 2*lnL
    BIC.append(bic)
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.plot(degrees, AIC, '-o',markersize=8, label='AIC')
ax.plot(degrees, BIC, '-o',markersize=8, label='BIC')
ax.set_xlabel('Polynomial Degree')
ax.set_xlim(.9, 5.1)
prt.prettyfy(ax, xmaj=1, xmin=1, ymaj=10, ymin=2, xfmt='%d',yfmt='%d')
ax.legend()
plt.show()
```

The AIC and BIC do not decrease significantly beyond a polynomial fit of degree 2. A higher order fit is not justified by the data.

```python
polynomial = PolynomialRegression(2)
polynomial.fit(x[:,None], medgr)
fit = polynomial.predict(x[:,None])
```

```python
disk = dfGold[(vtan < 150)]
M_r_disk = Mr_est[np.where(vtan < 150)]
DM_disk = disk.rObs-M_r_disk

halo = dfGold[(vtan > 200)]
M_r_halo = Mr_est[np.where(vtan > 200)]
DM_halo = halo.rObs-M_r_halo

DMbins = np.linspace(3,15,50)
M_rbins = np.linspace(10,18,50)
Mrvalues =  0.5 * (M_rbins[1:] + M_rbins[:-1])
DMvalues =  0.5 * (DMbins[1:] + DMbins[:-1])
distances = 10.**(DMvalues/5.+1.)
```

```python
histD, xbins, ybins = np.histogram2d(DM_disk, M_r_disk, bins=(DMbins, M_rbins))
histH, xbins, ybins = np.histogram2d(DM_halo, M_r_halo, bins=(DMbins, M_rbins))
```

```python
from astroML.lumfunc import binned_Cminus, bootstrap_Cminus, Cminus
import os

DMmax_disk = 24.5 - M_r_disk
M_rmax_disk = 24.5 - DM_disk
DMmax_halo = 24.5 - M_r_halo
M_rmax_halo = 24.5 - DM_halo

if not os.path.exists('Cminus.npz'):
    dist_DM_disk, err_DM_disk, dist_M_r_disk, err_M_r_disk = bootstrap_Cminus(DM_disk, M_r_disk, DMmax_disk, M_rmax_disk, DMbins, M_rbins, \
         Nbootstraps=20, normalize=True)
    dist_DM_halo, err_DM_halo, dist_M_r_halo, err_M_r_halo = bootstrap_Cminus(DM_halo, M_r_halo, DMmax_halo, M_rmax_halo, DMbins, M_rbins, \
         Nbootstraps=1, normalize=True)
    # I don't know why it breaks when I try to do more than one bootstrap on the halo
    
    np.savez('Cminus.npz', dist_DM_disk=dist_DM_disk, err_DM_disk=err_DM_disk, dist_M_r_disk=dist_M_r_disk, err_M_r_disk=err_M_r_disk,\
                           dist_DM_halo=dist_DM_halo, err_DM_halo=err_DM_halo, dist_M_r_halo=dist_M_r_halo, err_M_r_halo=err_M_r_halo)
else:
    data = np.load('Cminus.npz')
    dist_DM_disk=data['dist_DM_disk']
    err_DM_disk=data['err_DM_disk']
    dist_M_r_disk=data['dist_M_r_disk']
    err_M_r_disk=data['err_M_r_disk']
    
    dist_DM_halo=data['dist_DM_halo']
    err_DM_halo=data['err_DM_halo']
    dist_M_r_halo=data['dist_M_r_halo']
    err_M_r_halo=data['err_M_r_halo']
    
```

```python
# compute cumulative distribution functions
n_disk = [] ; phi_disk = []
n_halo = [] ; phi_halo = []

for ii in range(0,len(DMbins)-1):
    n_disk.append(np.trapz(dist_DM_disk[:ii], DMbins[:ii]))
    phi_disk.append(np.trapz(dist_M_r_disk[:ii], M_rbins[:ii]))
    n_halo.append(np.trapz(dist_DM_halo[:ii], DMbins[:ii]))
    phi_halo.append(np.trapz(dist_M_r_halo[:ii], M_rbins[:ii]))
    
n_disk = np.array(n_disk); phi_disk = np.array(phi_disk)
n_halo = np.array(n_halo); phi_halo = np.array(phi_halo)
```

```python
DM_0 = DMbins[28]
Mr_0 = M_rbins[25]

print 'DM_0:', DM_0
print 'Mr_0:', Mr_0

# compute number of stars in sample within given cuts
N_0_disk = DM_disk[(DM_disk < DM_0) & (M_r_disk < Mr_0)].shape[0]
N_0_halo = DM_halo[(DM_halo < DM_0) & (M_r_halo < Mr_0)].shape[0]

# Fraction of sky covered by sample
dOmg = 2.*np.pi*(1.-np.cos(np.radians(30)))
fsky = dOmg/(4.*np.pi)
# Compute normalization constant
C_disk= N_0_disk/(n_disk[28]*phi_disk[25])
C_halo= N_0_halo/(n_halo[28]*phi_halo[25])
print 'C_disk:', C_disk
print 'C_halo:', C_halo
```

```python
rho_disk = C_disk * 5/np.log(10)/dOmg / distances**3 * dist_DM_disk
rho_halo = C_halo * 5/np.log(10)/dOmg / distances**3 * dist_DM_halo

err_DM_disk = rho_disk * err_DM_disk / dist_DM_disk
err_DM_halo = rho_halo * err_DM_halo / dist_DM_halo
```

```python
m_disk, b_disk = np.polyfit(distances[(distances>100.)&(distances<500.)],np.log(rho_disk)[(distances>100.)&(distances<500.)],1)
rho0_disk = np.exp(b_disk)

rho0_halo = np.median(rho_halo[(distances>100.)&(distances<500.)])

normdist_Mr_disk = rho0_disk*dist_M_r_disk
normdist_Mr_halo = rho0_halo*dist_M_r_halo

err_M_r_disk = err_M_r_disk * rho0_disk
err_M_r_halo = err_M_r_halo * rho0_halo

print 'rho_0_disk:', rho0_disk
print 'rho_0_halo:', rho0_halo
```

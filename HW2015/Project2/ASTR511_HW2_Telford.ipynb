```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from astroML.stats import median_sigmaG
mpl.rc('xtick', labelsize=16)     
mpl.rc('ytick', labelsize=16)  
mpl.rc('axes', labelsize=20)
mpl.rc('legend', fontsize=20)
mpl.rc('lines', linewidth=2)
mpl.rc('font', size=18, family='serif', style='normal', variant='normal', stretch='normal', weight='normal')
%matplotlib inline
```

```python
# get data
data_sim = np.genfromtxt('LSSTsimWDtruth60.dat', names = ['ra','dec','u','g','r','i','z','y', \
                                                          'Mr','logg','vR','vPhi','vZ','T','P'], dtype=None)
data_obs = np.genfromtxt('LSSTsimWDobs60.dat', names = ['ra','dec','uObs','uErr','gObs','gErr', 
                                                        'rObs','rErr','iObs','iErr','zObs','zErr', \
                                                        'yObs','yErr','piObs','piErr','muRAObs','muDecObs','muErr'], dtype=None)
```

## Problem 1: Distribution in $\Delta$Distance Modulus and Color-Color Diagram for High S/N WDs

```python
high_sn = data_obs['piObs']/data_obs['piErr'] > 10.
gold_samp = data_obs[high_sn]

d_pc = 1000./gold_samp['piObs'] #distance in pc
dm_obs = 5. * np.log10(d_pc) - 5.

dm_true = data_sim['r'][high_sn] - data_sim['Mr'][high_sn]

dm_diff = dm_obs - dm_true
```

```python
# compute statistics

med, sigma_g = median_sigmaG(dm_diff)
avg = np.mean(dm_diff)
RMSE = np.std(dm_diff)

print 'Median: ', med
print 'Mean: ', avg
print 'RMSE: ', RMSE
print 'Sigma_G: ', sigma_g

# clip outliers
dm_diff = dm_diff[np.abs(dm_diff - med) < 3 * sigma_g]

# plot the difference distribution with Knuth rule binning
hist, bin_edges = np.histogram(dm_diff, bins=20)

# for checking that bin size agrees with Knuth rule
#print 2.7 * sigma_g / float(len(dm_diff))**(1./4)
#print bin_edges[1] - bin_edges[0]

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(bin_edges[1:], hist, linestyle='steps')
ax.set_xlabel(r'$\Delta \mu$')
ax.set_ylabel(r'$\mathrm{d}N/\mathrm{d}\Delta \mu$')
```

These errors seem reasonable to me; most of the errors are less than 0.2 mag, so roughly 20%. The distribution is not Gaussian, since the RMSE and $\sigma_G$ are not equal. Further, the plot below shows that a Gaussian fit does not describe the distribution very well. I wouldn't necessarily expect the errors to be Gaussian-distributed.

This sample covers an area of 0.812 steradians, or about 2763.4 square degrees, so there should be about 7.2 times more WDs in the full LSST survey. For a "gold proper motion" sample in the full survey, there should be about 51523 WDs.

```python
#fit gaussian distribution and overplot
hist, bin_edges = np.histogram(dm_diff, bins=20, normed=True)

from scipy.stats import norm
mu, std = norm.fit(dm_diff)
x = np.linspace(bin_edges[1],bin_edges[-1],100)
gaussian = norm.pdf(x, mu, std)

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(bin_edges[1:], hist, linestyle='steps')
ax.plot(x, gaussian, 'r-')
ax.set_xlabel(r'$\Delta \mu$')
ax.set_ylabel(r'$\mathrm{d}N/\mathrm{d}\Delta \mu$')
```

```python
# plot color-color diagram
ug = gold_samp['uObs'] - gold_samp['gObs']
gr = gold_samp['gObs'] - gold_samp['rObs']

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(ug, gr, '.', ms=0.8)
ax.set_xlabel(r'$g-r$')
ax.set_ylabel(r'$u-g$')
```

This distribution does look much crisper than that in the Ivezic et al. paper, demonstrating that repeated observations with LSST will be able to significantly increase the S/N of such observations over what is currently achievable.

## Problem 2: Fitting Photometric Parallax Relation for H-Rich WDs

```python
# calculate Mr_obs; select H rich WDs; plot color-magnitude diagram and median values
H_population = data_sim['T'][high_sn] == 1
Mr_obs = gold_samp['rObs'] - dm_obs

Mr_H = Mr_obs[H_population]
gr_H = gr[H_population]

gr_bin_edges = np.arange(-0.5, 1.1, 0.1)
gr_bins = np.digitize(gr_H, gr_bin_edges)
Mr_medians = []
gr_medians = []
for index in np.arange(1, len(gr_bin_edges)+1):
    Mr_medians.append(np.median(Mr_H[gr_bins == index]))
    gr_medians.append(np.median(gr_H[gr_bins == index]))

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(gr_H, Mr_H, '.', ms=0.5)
ax.plot(gr_medians, Mr_medians, 'ro', ms=5)
ax.set_xlabel(r'$g-r$')
ax.set_ylabel(r'$M_r$')
ax.set_xlim([-0.5, 1.5])
```

```python
# fit relation to median values
from sklearn.metrics import mean_squared_error

fig, ax = plt.subplots(figsize=(8,6))
#ax.plot(gr_H, Mr_H, '.', ms=0.5)
ax.plot(gr_medians, Mr_medians, 'o', ms=5)

for p in np.arange(1,4):
    fit = np.polyfit(gr_medians, Mr_medians, p)
    print fit
    f = np.poly1d(fit)
    rms = np.sqrt(mean_squared_error(Mr_medians, f(gr_medians)))
    ax.plot(gr_medians, f(gr_medians), '--', label='Degree '+str(p)+': RMSE = '+str(rms))
    
ax.set_xlabel(r'$g-r$')
ax.set_ylabel(r'$M_r$')
ax.set_xlim([-0.5, 1.5])
plt.legend(loc=4, fontsize = 14)
```

I selected a second degree polynomial as the fitting function because going from a linear to quadratic fit gave a significant decrease in the RMSE (and improved the fit qualitatively), but going to a cubic fit did not cause the RMSE to decrease substantially. It is prefereable to choose the lowest order polynomial that gives a good fit in order to minimize variance. The second order best fit is $M_r = 12.60 + 4.18(g-r) - 0.65(g-r)^2$. This relation is only valid over the range in which it was calibrated, so $-0.4 < g-r < 1.0$.

## Problem 3: Completeness and Contamination for Varying $v_{tan}^{cut}$ to Define Disk/Halo Samples

```python
# gold proper motion sample
r_cut = data_obs['rObs'] < 24.5
pm_samp = data_obs[r_cut]

pm_tot = np.sqrt(pm_samp['muRAObs']**2 + pm_samp['muDecObs']**2)
pm_SNR = pm_tot / pm_samp['muErr']

frac = np.sum(pm_SNR > 3) / float(np.shape(pm_samp)[0])

print 'The fraction of this sample with proper motion SNR > 3 is: ', frac
```

```python
# apply relation to dermine Mr; calculate distance and v_tan
gr_pm = pm_samp['gObs'] - pm_samp['rObs']

Mr_pm = 12.6 + 4.18 * gr_pm - 0.65 * gr_pm **2
d_pm = 10**((pm_samp['rObs'] - Mr_pm) / 5. + 1.) #in parsecs

v_t = 4.74 * pm_tot * d_pm / 1000. #gives tangential velocity in km/s; pm_tot is in mas/yr

pm_samp_true = data_sim[r_cut]
disk_true = pm_samp_true['P'] == 1 #array of booleans indicating true member of disk population
halo_true = pm_samp_true['P'] == 2

disk_completeness = []
disk_contamination = []
halo_completeness = []
halo_contamination = []
v_cuts = np.arange(0., 500., 20.)

for v_cut in v_cuts:
    disk_samp = v_t < v_cut #array of booleans indicating selection as part of disk sample
    halo_samp = v_t > v_cut
    correct_disk_samp = disk_samp * disk_true
    correct_halo_samp = halo_samp * halo_true
    
    disk_completeness.append(np.sum(correct_disk_samp)/float(np.sum(disk_true)))
    disk_contamination.append(np.sum(disk_samp * halo_true)/float(np.sum(disk_samp)))
    halo_completeness.append(np.sum(correct_halo_samp)/float(np.sum(halo_true)))
    halo_contamination.append(np.sum(halo_samp * disk_true)/float(np.sum(halo_samp)))
    
fig, ax = plt.subplots(1,2, figsize=(12,6))

ax[0].plot(v_cuts, disk_completeness, 'o-', ms=5, label='Disk')
ax[0].plot(v_cuts, halo_completeness, 'o-', ms=5, label='Halo')
ax[0].set_xlabel(r'$v_{cut} (\mathrm{km/s})$')
ax[0].set_ylabel('Completeness')
ax[0].legend(loc=7, fontsize=14)

ax[1].plot(v_cuts, disk_contamination, 'o-', ms=5, label='Disk')
ax[1].plot(v_cuts, halo_contamination, 'o-', ms=5, label='Halo')
ax[1].set_xlabel(r'$v_{cut} (\mathrm{km/s})$')
ax[1].set_ylabel('Contamination')
ax[1].legend(loc=7, fontsize=14)

plt.tight_layout()
```

## Problem 4: The WD Luminosity Function for the Disk and Halo

```python
disk_sample = pm_samp[v_t < 150] 
halo_sample = pm_samp[v_t > 200]

disk_Mr = Mr_pm[v_t < 150] 
halo_Mr = Mr_pm[v_t > 200]

disk_ds = d_pm[v_t < 150] 
halo_ds = d_pm[v_t > 200]

disk_mus = disk_sample['rObs'] - disk_Mr
halo_mus = halo_sample['rObs'] - halo_Mr

mr_max = np.max(pm_samp['rObs'])
```

Given the plots in the previous problem, the completeness is almost 100% (slightly less) for the disk sample and roughly 60% for the halo sample; the contamination is just a few % for both the disk and halo samples. 

```python
# look at 2D distribution for halo sample
fig, ax = plt.subplots(figsize=(8,6))
H, xbins, ybins = np.histogram2d(halo_ds, halo_Mr, bins=(np.linspace(5., 6175., 51), np.linspace(10.4, 17.4, 51)))
ax.imshow(H.T, origin='lower', aspect='auto',
          interpolation='nearest', cmap=plt.cm.binary,
          extent=(xbins[0], xbins[-1], ybins[0], ybins[-1]))

# plot the cutoff curve
drange = np.linspace(5., 6175., 100)
Mmax = mr_max - (5 * np.log10(drange) - 5)
ax.plot(drange, Mmax, '-k')
ax.set_xlabel('Distance (pc)')
ax.set_ylabel(r'$M_r$')

#Mr_max = mr_max - (halo_sample['rObs'] - halo_Mr) # sanity check
#ax.plot(halo_ds, Mr_max, 'r.')
```

```python
# run Cminus method to get true distributions normalized to 1
from astroML.lumfunc import bootstrap_Cminus

# disk first
M_bins = np.linspace(10., 17., 40)  
mu_bins = np.linspace(0., 15., 40)

mr = disk_sample['rObs']
Mr_max = mr_max - (mr - disk_Mr)
mu_max = mr_max - disk_Mr

dist_mu_disk, err_mu_disk, dist_M_disk, err_M_disk = bootstrap_Cminus(disk_mus, disk_Mr, mu_max, Mr_max, mu_bins, M_bins,
                                                    Nbootstraps=5, normalize=True)

# now halo 
mr = halo_sample['rObs']
Mr_max = mr_max - (mr - halo_Mr)
mu_max = mr_max - halo_Mr

dist_mu_halo, err_mu_halo, dist_M_halo, err_M_halo = bootstrap_Cminus(halo_mus, halo_Mr, mu_max, Mr_max, mu_bins, M_bins,
                                                    Nbootstraps=3, normalize=True)
```

```python
# look at 2D distribution for both disk and halo
fig, ax = plt.subplots(1,2, figsize=(12,4))
ax[0].hexbin(halo_sample['rObs'] - halo_Mr, halo_Mr, cmap=plt.cm.binary)
ax[0].set_xlabel(r'$\mu$')
ax[0].set_ylabel(r'$M_r$')
ax[0].set_title('Halo', fontsize=18)
ax[0].set_xlim([4,14])
ax[0].set_ylim([11,16])
ax[0].invert_yaxis()


ax[1].hexbin(disk_sample['rObs'] - disk_Mr, disk_Mr, cmap=plt.cm.binary)
ax[1].set_xlabel(r'$\mu$')
ax[1].set_ylabel(r'$M_r$')
ax[1].set_title('Disk', fontsize=18)
ax[1].set_xlim([4,14])
ax[1].set_ylim([11,16])
ax[1].invert_yaxis()

plt.tight_layout()
print len(halo_Mr), len(disk_Mr)
```

```python
# define volume limited samples for disk and halo and count the number in each
vol_lim_disk = (disk_Mr < 13.5) * (disk_mus < 11.)
disk_vol_lim_count = np.sum(vol_lim_disk)

vol_lim_halo = (halo_Mr < 13.5) * (halo_mus < 11.)
halo_vol_lim_count = np.sum(vol_lim_halo)

print 'Disk: ', disk_vol_lim_count, ', Halo: ', halo_vol_lim_count
```

```python
# check that distributions all integrate to 1
from scipy.integrate import trapz

print 'Halo M: ', trapz(dist_M_halo, M_bins[1:])
print 'Disk M: ', trapz(dist_M_disk, M_bins[1:])
print 'Halo mu: ', trapz(dist_mu_halo, mu_bins[1:])
print 'Disk mu: ', trapz(dist_mu_disk, mu_bins[1:])
```

```python
# integrate each distribution up to the cutoff value to get fractions of distributions in volume limited samples
M_lim = M_bins < 13.5
mu_lim = mu_bins < 11.

frac_M_halo = trapz(dist_M_halo[M_lim], M_bins[M_lim])
frac_M_disk = trapz(dist_M_disk[M_lim], M_bins[M_lim])
frac_mu_halo = trapz(dist_mu_halo[mu_lim], mu_bins[mu_lim])
frac_mu_disk = trapz(dist_mu_disk[mu_lim], mu_bins[mu_lim])

print frac_M_halo, frac_M_disk, frac_mu_halo, frac_mu_disk
```

```python
# get normalization for disk and halo
norm_disk = disk_vol_lim_count / (frac_M_disk * frac_mu_disk)
norm_halo = halo_vol_lim_count / (frac_M_halo * frac_mu_halo)

print 'Disk normalization: ', norm_disk
print 'Halo normalization: ', norm_halo
```

```python
# calculate the stellar density as a function of distance accounting for fraction of area on sky covered
dOmega = 2. * np.pi * (1. - np.cos(np.radians(30.)))

mu_mid = 0.5 * (mu_bins[1:] + mu_bins[:-1])
Dpc = 10 * 10**(0.2 * mu_mid)

rho_disk = norm_disk * 5/np.log(10)/dOmega / Dpc**3 * dist_mu_disk
rho_halo = norm_halo * 5/np.log(10)/dOmega / Dpc**3 * dist_mu_halo

rho_error_disk = err_mu_disk * rho_disk / dist_mu_disk
rho_error_halo = err_mu_halo * rho_halo / dist_mu_halo

# fit inner part of stellar density distribution to estimate value in stellar neighborhood (distance modulus 0)
m_disk, b_disk = np.polyfit(Dpc[(Dpc>100.)&(Dpc<500.)],np.log(rho_disk)[(Dpc>100.)&(Dpc<500.)],1)
rho0_disk = np.exp(b_disk)

rho0_halo = np.median(rho_halo[(Dpc>100.)&(Dpc<500.)])

# renormalize the M distributions
M_disk_norm = dist_M_disk * rho0_disk
M_halo_norm = dist_M_halo * rho0_halo

M_error_disk = err_M_disk * rho0_disk
M_error_halo = err_M_halo * rho0_halo
```

```python
# read in true luminosity functions for comparison
trueDiskLF = np.genfromtxt('WDlumfuncDisk.dat', names=['Mr', 'LF'], dtype=None)
trueHaloLF = np.genfromtxt('WDlumfuncHalo.dat', names=['Mr', 'LF'], dtype=None)

# plot the calculated and true LFs
fig, ax = plt.subplots(figsize=(8,6))

M_mid = 0.5 * (M_bins[1:] + M_bins[:-1])

ax.errorbar(M_mid, M_disk_norm, M_error_disk, fmt='bo', ms=5, lw=1, label='Disk')
ax.errorbar(M_mid, M_halo_norm, M_error_halo, fmt='go', ms=5, lw=1, label='Halo')

ax.plot(trueDiskLF['Mr'], trueDiskLF['LF'], 'b-', label='True Disk')
ax.plot(trueHaloLF['Mr'], trueHaloLF['LF']/200., 'g-', label='True Halo')
ax.set_xlabel(r'$M_r$')
ax.set_ylabel(r'$\mathrm{d}^2 N/\mathrm{d}V \mathrm{d}M_r$')
ax.set_yscale('log')
ax.legend(loc=4, fontsize=14)
```

Overall, agreement is quite good between the true and observed LFs. The observed disk LF disagrees with the true only at the very faint end, which isn't shocking.

## Problem 5: WD Spatial Distribution for the Disk and Halo

```python
# plot stellar density as a function of distance
fig, ax = plt.subplots(figsize=(8,6))

ax.errorbar(Dpc, rho_disk, rho_error_disk, fmt='bo', ms=5, lw=1, label='Disk')
ax.errorbar(Dpc, rho_halo, rho_error_halo, fmt='go', ms=5, lw=1, label='Halo')

ax.set_xlabel(r'$d$ (pc)')
ax.set_ylabel(r'$\mathrm{d} \rho/\mathrm{d}d$')
ax.set_yscale('log')
ax.set_xlim([0,5000])
ax.legend(loc=1, fontsize=14)
```

The distribution in the previous HW looked like the disk distribution here, which makes sense because the halo stellar density is so much smaller than the disk stellar density at small distances that the disk distribution will dominate.

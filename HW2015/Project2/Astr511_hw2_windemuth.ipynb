```python
from astroML.stats import median_sigmaG

def knuth(x, retN=True):
    binsize = 2.7 * 0.7413 * (np.percentile(x, 75)-np.percentile(x, 25)) / (x.size**(1./4.))
    Nx = int(np.ceil((x.max()-x.min())/binsize))
    if retN:
        return binsize, Nx
    return binsize

def getdiff(x, xtrue, nbins=None, rms=False):
    diff = x-xtrue
    if nbins is None:
#        knuth = 2.7*diff.std() / (len(diff)**(1./3.))
#        Npts = int(np.ceil((diff.max()-diff.min())/knuth))
        binsize, Npts = knuth(diff, retN=True)
        #nbins = np.arange(diff.min(), Npts*binsize+diff.min(), binsize)
        nbins = np.linspace(diff.min(), diff.max(), Npts, endpoint=True)
    #print nbins, diff.min(), diff.max()
    diff_count, diff_edges = np.histogram(diff, nbins)
    if rms:
        #clip = (abs(diff-np.median(diff)) < 3*np.std(diff))
        #return (np.std(diff), np.median(diff[clip]), np.mean(diff[clip]**2)**0.5)
        diff_med, diff_sigG = median_sigmaG(diff)
        return np.mean(diff), np.std(diff), diff_med, diff_sigG
    return ((diff_edges[1:]+diff_edges[:-1])/2., diff_count)
```

## B. ##
Using the “gold parallax sample” from A, estimate the absolute $r$ band magnitude as Mobs=rObs−DMobs, with the observed distance modulus, DMobs, determined using the “measured” trigonometric parallax, piObs. Plot Mobs vs. (gObs−rObs) color for stars withT=1 (i.e., hydrogen white WDs). Fit a low-order polynomial to derive a photometric parallax relation, $M_r(g−r)$ (hint: you may want to first compute the median $M_r$ in about 0.1 mag wide bins of the $g−r$ color, and then fit a polynomial to these median values vs. $g−r$ bin value). Howdid you choose the order of your polynomial fit? In what range of $M_r$ and $(g−r)$ is your relation valid?

```python
mobs = obs[:, ohead['r']] - odmod
cut = gold * (truth[:, thead['T']]==1)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)
ax_in = fig.add_axes([0.55, 0.2, 0.3, 0.35])

ax.plot(gr[cut],mobs[cut], 'k.', alpha=0.1)
ax.set_xlabel('observed g-r')
ax.set_ylabel('observed M$_r$')

def binquant(x, y, binsize=0.1, retbin=False):
    nbins = int(np.ceil((x.max()-x.min())/binsize))
    nbins = x.min() + np.arange(nbins)*binsize
    inds = np.digitize(x, nbins)
    #avgx=(nbins[1:]+nbins[:-1])/2.
    avgx = np.zeros(nbins.size-1)
    avgy = np.zeros(avgx.size)
    for ii in range(nbins.size-1):
        good = np.where(inds == ii+1)[0]
        avgx[ii] = np.mean(x[good])
        avgy[ii] = np.mean(y[good])
    if retbin:
        return (nbins[1:]+nbins[:-1])/2., avgy
    return avgx, avgy

avgx, avgy = binquant(gr[cut],mobs[cut])

for ii in range(1, 8):
    coeff = np.polyfit(avgx, avgy, ii)
    linfit = np.poly1d(coeff)
    ax.plot(avgx, linfit(avgx), '-', label='order '+str(ii))
    bic = np.sum(((linfit(gr[cut])-mobs[cut])/(egr[cut]))**2) + (ii+1)*np.log(len(gr[cut]))
    if ii==1:
        bic0 = bic.copy()
    ax_in.plot(ii, bic/bic0, 'o-')
ax.plot(avgx, avgy, 'yo')
ax.legend(loc=2)
ax_in.set_xlabel('Polynomial order')
ax_in.set_ylabel('Normalized $\chi^2$')
ax_in.set_title('BIC Distribution')
ax.set_xlim(-0.6, 1.3)
ax.set_ylim(9.8, 16.5)
best_coeff = np.polyfit(avgx, avgy, 3)
print "3rd order: ", best_coeff
```

I used a Bayesian Information Criterion test to see the $\chi^2$ improvement as a function of polynomial order. The BIC distribution starts to flatten at a 3rd order polynomial (meaning beyond this point we would be overfitting the data). So a $\sim$ 3rd order polynomial is optimal:

$y = 0.67858x^3  -1.21783x^2  + 4.08603x^1 +  12.6701571x^0$ ,

where $y = M_r$ and  $x = (g-r)$. Note this is valid for $-0.5<g-r<1.1$.

```python
import os.path

def getmax(x, offset, retdist=True):
    return offset - x

def getmax2(x, offset, retdist=True):
    if retdist:
        return 10.*10**((offset-x)/5.)
    #print rmax
    return offset - 5.*np.log10(x/10.)

def computeLF(x, y, offset, archive_file, rectedges=None, Nbootstraps=20):
    if not os.path.exists(str(archive_file)):
        print "Computing bootstrapped luminosity function"
        xmax = getmax(y, offset, retdist=True)
        ymax = getmax(x, offset, retdist=False)
        if rectedges is None:
            xbinsize, Nx = knuth(x, retN=True)
            ybinsize, Ny = knuth(y, retN=True)
            xbins = np.linspace(x.min(), x.max(), Nx+1, endpoint=True)
            ybins = np.linspace(y.min(), y.max(), Ny+1, endpoint=True)
            xbins = np.linspace(x.min(), x.max(), 40)
            ybins = np.linspace(y.min(), y.max(), 40)
        else:
            xbinsize, Nx = knuth(x[(x<rectedges[0])], retN=True)
            ybinsize, Ny = knuth(y[(y<rectedges[1])], retN=True)
            xbins = np.linspace(x.min(), x[(x<rectedges[0])].max(), Nx+1, endpoint=True)
            ybins = np.linspace(y.min(), y[(y<rectedges[1])].max(), Ny+1, endpoint=True)
            xbins = np.linspace(x.min(), x[(x<rectedges[0])].max(), 40)
            ybins = np.linspace(y.min(), y[(y<rectedges[1])].max(), 40)
        #xbins = xbins[xbins<=5000.]
        #xbins, ybins = np.linspace(0, 7000., 42), np.linspace(10, 18, 42)
        dist_x, err_x, dist_y, err_y = bootstrap_Cminus(x, y, xmax, ymax, xbins, ybins, \
                                                        Nbootstraps=Nbootstraps, normalize=True)
        np.savez(archive_file, xbins=xbins, dist_x=dist_x, err_x=err_x,
                 ybins=ybins, dist_y=dist_y, err_y=err_y, xmax=xmax, ymax=ymax)
    else:
        print "Loading bootstrapped LF from archive file"
        archive = np.load(archive_file)
        dist_x = archive['dist_x']
        dist_y = archive['dist_y']
        err_x, err_y = archive['err_x'], archive['err_y']
        xbins, ybins = archive['xbins'], archive['ybins']
        xmax, ymax = archive['xmax'], archive['ymax']
    return dist_x, err_x, dist_y, err_y, xbins, ybins, xmax, ymax
```

```python
def makecen(val):
    return 0.5*(val[1:]+val[:-1])

def normplot(xx, yy, x, y, ex, ey, xbins, ybins, rect=[11., 13.5], disk=True):
    if disk:
        title, color ='Disk', 'b'
        trueLF = diskLF
        norm = 1.# 274193.360759 + 148664.825157 #np.sum(diskcand)+np.sum(halocand)
    else:
        title, color ='Halo', 'r'
        trueLF = haloLF
        norm = 1. #274193.360759 + 148664.825157 

    xcen = makecen(xbins)
    ycen = makecen(ybins)
    xcenN = xcen[np.argsort(x)][-1]

#    H, xedges, yedges = np.histogram2d(xx, yy, bins=(xbins, ybins), 
#                                   normed=False)

#    Harea_all = np.trapz(H, xcen, axis=0)
#    Harea_all = np.trapz(Harea_all, ycen)
#    Harea = np.trapz(H[xcen<=rect[0], :], xcen[xcen<=rect[0]], axis=0)
#    Harea = np.trapz(Harea[ycen<=rect[1]], ycen[ycen<=rect[1]])

#    const = Harea / (np.trapz(x[xcen<=rect[0]], xcen[xcen<=rect[0]]) * \
#                 np.trapz(y[ycen<=rect[1]], ycen[ycen<=rect[1]]))
    Nmask = (xx < rect[0]) * (yy < rect[1])
#    const = np.sum(Nmask) / (np.trapz(x[xcen < rect[0]], xcen[xcen < rect[0]]) * \
#                 np.trapz(y[ycen < rect[1]], ycen[ycen < rect[1]]))
    const = np.sum(Nmask) / (np.sum((x*(xbins[1:]-xbins[:-1]))[xcen <= rect[0]]) * \
                             np.sum((y*(ybins[1:]-ybins[:-1]))[ycen <= rect[1]]))


    Dpc = 10 * 10**(0.2 * xcen)
#    fsky = .8418 #/(2*np.pi)#(15./90.)**2
#    csky = const / fsky
    domega = 2.*np.pi*(1.-np.cos(np.pi/180. * (90-60.)))
#    print domega
    rho = const * 5/np.log(10)/domega / Dpc**3 * x 
    erho = rho * ex / x
#    indx0 = np.searchsorted(xcen, rect[0])
#    indxN = np.searchsorted(xcen, xcenN)
    #rho0 = np.interp(rect[0], xcen[indx0-1:indx0+1],rho[indx0-1:indx0+1])
    if disk:
        rho_m, rho_b = np.polyfit(Dpc[(Dpc>150) * (Dpc < 500)], 
                                  np.log10(rho[(Dpc>150) * (Dpc < 500)]), 1)
        rho0 = 10**rho_b
    else:
        rho0 = np.median(rho[(Dpc>500) * (Dpc < 2000)])
    #rho0 = np.interp(10, Dpc[5:10], rho[5:10])
#    rhoN = np.interp(10 * 10**(0.2*xcenN), Dpc[indxN-1:indxN+1],rho[indxN-1:indxN+1])
    print title
    print "observed sample size, size of volume-limited sample, C, rho0 =", \
        xx.size, np.sum(Nmask), const, rho0
    #print xcenN, rho0, rhoN
    #print xcen[indx0-1:indx0+1],rho[indx0-1:indx0+1], xcen[indxN-1:indxN+1],rho[indxN-1:indxN+1]
    psi = y * rho0#/rhoN
    epsi = psi * ey / y
    
    plt.figure(figsize=(12,6))
    plt.subplot(121)
    plt.plot(trueLF[:, 0], trueLF[:, 1], color+'-', lw=1, label='true')
    plt.errorbar(ycen, psi, epsi, fmt = color+'^', label='model')
    plt.xlabel('$M_r$ (mag)')
    plt.ylabel('$\psi(M_r)$ mag$^{-1}$ pc$^{-3}$')
    plt.yscale('log')
    plt.title(title)
    plt.legend(loc =4)
    plt.subplot(122)
    plt.errorbar(Dpc/1000., (rho), erho, fmt = color+'^', label='observed rho')
    plt.plot(0./1000., (rho0), color+'o', label='extrap to 0')
#    plt.plot(10*10**(0.2*xcenN)/1000., (rhoN), color+'d', label='interp to N')
    plt.yscale('log')
    plt.legend()
    plt.xlabel('d (kpc)')
    plt.ylabel('$\\rho$')
    plt.xlim(-0.1, 6)
    return (Dpc, rho, erho)

disk_spatial = normplot(mu[diskcand], Mr[diskcand], disk_x, disk_y, disk_ex, 
                 disk_ey, disk_xbins, disk_ybins)
halo_spatial = normplot(mu[halocand], Mr[halocand], halo_x, halo_y, halo_ex, 
                  halo_ey, halo_xbins, halo_ybins, disk=False)
```

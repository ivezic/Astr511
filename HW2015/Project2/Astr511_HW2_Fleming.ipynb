Below, define useful histogram and plotting functions.  The histogram function uses Knuth's Rule to bin data in order to best show the shape of the distribution.

```python
#Useful histogram plotting function

def plotHist(data,mask=None):
    """
    Given a data set and an optional mask to select a data subset, return data used to plot a histrogram.
    Computes optimal bin size via Knuth's Rule: binSize = 2.7*sigma/(N^1/4) where sigma = 0.7413*(q_75 - q_25)
    
    Input:
    Data:1D data set in question
    Mask: Means to select subset of data if interested, defaults to true array for selecting all data
    
    Output:
    x, y arrays to use for plotting a histrogram
    """
    
    #Select subset
    if mask != None:
        data = data[mask]
    
    #Compute bin size via Knuth's Rule
    
    #Compute 75th, 25th quartiles
    q_75 = np.percentile(data, 75)
    q_25 = np.percentile(data, 25)
    
    #Compute sigma
    sigma = 0.7413*(q_75-q_25)
    
    binSize = (2.7*sigma)/np.power(len(data),0.25)
    
    #Histrogram Data
    bins = np.floor((data.max() - data.min())/binSize)
    hist, edges = np.histogram(data,bins)
    x = (edges[:-1] + edges[1:])/2.0
    
    #Cut off 3sigma tails
    #Filter out data that's not within 3 sigma
    high = np.percentile(x, 99.85)
    low = np.percentile(x, 0.15) 
    mask = np.logical_and(x < high, x > low)
    hist = hist[mask]
    x = x[mask]
    
    #Compute median, mean
    mean = np.mean(data)
    median = np.median(data)
    
    return x, hist, sigma, median, mean
```

#Apply your photometric parallax relation from B to estimate $M_r$ and distance (using $M_r$ and rObs)

```python
#Define a function to compute the photometric parallax derived above
def photopar(x,coef):
    """
    Returns M_r given x, the g-r color
    """
    return coef[0]*x*x*x + coef[1]*x*x + coef[2]*x + coef[3]
```

```python
#Estimate Mr from golden proper motion sample
goldMuGR = goldMu[:,gObs] - goldMu[:,rObs]
goldMr = photopar(goldMuGR[:],coef)

#Define distance modulus for this subset
goldDM = goldMu[:,rObs] - goldMr

#Estimate distances
goldMuDistance = np.power(10,(goldDM/5.0))*10
```

#Use this distance to compute tangential velocity, $v_{tan}$ (of course, you also need the observed proper motion; be careful about units!).
Units Note: My goldMuDitance sample is in units of pc.  To Compute $v_{tan}$, apply the following formula: 
$v_{tan} = 4.74 \mu d$ where d is in pc, proper motion mu is in arcsec/yr and $v_{tan}$ is in km/s.  First, I need to compute the proper motions of the sample.  The proper motion is given by : $\mu^2 = \mu^2_{\delta} + \mu^2_{\alpha}\cos^2(\delta)$

```python
#Compute proper motion for samples in golden proper motion sample
#Divide by 1000 to put it in arcsec/yr instead of milliarcsec/yr
goldMuMu = np.sqrt(np.power(goldMu[:,muRAObs],2)+np.power(goldMu[:,muDecObs],2)*np.power(np.cos(goldMu[:,dec]),2))/1000

#Compute v_tan for the golden proper motion sample
goldMuVTan = 4.74*(goldMuMu)*goldMuDistance
```

The completeness is defined as the number of (disk,halo) objects in the selected subsample divided by the total number of such objects, and contamination is the number of objects of the “wrong” type in the selected subsample divided by the total number in that subsample.

Define useful functions for computing the completeness and contamination.

```python
#Define function to calculate halo, disk completion
def cutoffCompFraction(dataObs,dataTrue,trueMask,cutoff):
    """
    Function takes in a dataset and cutoff and returns the completeness of disk and halo objects.
    dataObs: observed subset data to slice depending on disk-halo cutoff. Data is v_tan.
    dataTrue: true data set
    trueMask: mask that selects disk WD from dataTrue set
    cutoff: Cutoff between disk/halo stars for dataObs
    """
    #Find number of true disk, halo stars
    trueDiskNum = len(dataTrue[trueMask])
    trueHaloNum = len(dataTrue[~trueMask])
            
    #define cutoff mask corresponding to cutoff
    cutoffMask = dataObs < cutoff
                
    #Find number of true disk, halo stars given cutoff
    cutOffDisk = np.sum(np.logical_and(cutoffMask,trueMask))
    cutOffHalo = np.sum(np.logical_and(~cutoffMask,~trueMask))
        
    return float(cutOffDisk)/trueDiskNum, float(cutOffHalo)/trueHaloNum

#end function

#Define function to calculate halo, disk contamination
def cutoffContFraction(dataObs,dataTrue,trueMask,cutoff):
    """
    Function takes in a dataset and cutoff and returns the contamination of disk and halo objects.
    dataObs: observed subset data to slice depending on disk-halo cutoff. Data is v_tan.
    dataTrue: true data set
    trueMask: mask that selects disk WD from dataTrue set
    cutoff: Cutoff between disk/halo stars for dataObs
    """
    #define cutoff mask corresponding to cutoff
    cutoffMask = dataObs < cutoff
                
    #Find number of disk, halo stars given cutoff
    cutOffDisk = float(np.sum(cutoffMask))
    cutOffHalo = float(np.sum(~cutoffMask))
                
    #Find how many of wrong type are in cutoff
    wrongDisk = float(np.sum(np.logical_and(cutoffMask,~trueMask)))
    wrongHalo = float(np.sum(np.logical_and(~cutoffMask,trueMask)))
    
    #Ensure no 1/0 error
    if cutOffDisk > 0.001:
        return wrongDisk/cutOffDisk, wrongHalo/cutOffHalo
    else:
        return 0.0, wrongHalo/cutOffHalo

#end function
```

#Normalize the distributions.

In order to normalize the luminosity functions, I select a region such that M < M0 && D < D0 for some M0 and D0 that are within my selection volume and immune to selection effects.  I can compute a constant C via the following equation:

$N^*(M < M^*$ and $D < D^* = C \Phi_c (M^*)n_c(D^*)$ where $M^*=M0$ and $D^*=D0$.  

$N^*$ is simply the number of points in this region.  To compute $\Phi_c$ and $n_c$, I create a function using scipy's linear interpolation method.  Next, I must normalize $\Phi_c$ and $n_c$ to 1 for the maximum values of M and D.  From these, I can estimate C.

Once I have an estimate for C, I can compute the number of points per unit 2D area:

$d^2N/(dMdD)= C(d\Phi_c(M)/DM)(dn_c(D)/dD)$.  

The derivatives of $\Phi_c$ and $n_c$ are computing using numpy's gradient function.

```python
#Calculate angular size of region on size we're looking at
bMin = 60.0 #min galactic lat
dOmega = 2.0*math.radians(180)*(1.0-math.cos(math.radians(90-bMin)))
```

```python
#Find values in sample that correspond to area defined by MrMax and DMMax
DMmax = 11.0
MrMax = r_max - DMmax
NmaskD = np.logical_and((cutDMDisk < DMmax),(cutMrDisk < MrMax)) 
Ndisk = np.sum(NmaskD)
NmaskH = np.logical_and((cutDMHalo < DMmax),(cutMrHalo < MrMax)) 
Nhalo = np.sum(NmaskH)
print 'Size of volume-limited samples D and H:', Ndisk, ' and ', Nhalo,  '(Dist<', DMmax, ', Mr<', MrMax, ')'
```

```python
#Define helper function from Ivezic
def IntDiffDistr(x, y, xLimit):
    ## return the integral of y(x) up to xmax (from the starting point)
    ## it is assumed that the length of x is by 1 larger than the length of y
    ## x represents bin boundaries for a piecewise constant y(x)
    # integrated function (the term in parenthesis is dx)
    intfunc = y * (x[1:] - x[:-1])
    xmax = x[1:]
    return np.sum(intfunc[xmax <= xLimit])
```

```python
#Compute fraction of total sample contained in simple region 
XfracD = IntDiffDistr(DMDBins,dist_DDM,DMmax)
YfracD = IntDiffDistr(MrDBins, dist_DMr,MrMax)
XfracH = IntDiffDistr(DMHBins,dist_HDM,DMmax)
YfracH = IntDiffDistr(MrHBins,dist_HMr,MrMax)
```

```python
#Compute, output Cdisk, Chalo using fraction of samples in simple region
Cdisk = Ndisk/(XfracD*YfracD)
Chalo = Nhalo/(XfracH*YfracH)
print 'Cdisk = ', Cdisk 
print 'Chalo = ', Chalo
```

```python
#Number density normalization

#Compute midpoints for plotting/calculations
x_mid = 0.5 * (DMDBins[1:] + DMDBins[:-1])
y_mid = 0.5 * (MrDBins[1:] + MrDBins[:-1])

#Convert from DM to pc^-1:  dDM/dV = 5 / [ln(10) * dOmega * D^3] 
Dpc = 10 * 10**(0.2*x_mid)

#Units are "number of stars per pc^3" (of all absolute magnitudes Mr) 
rhoD = Cdisk * 5/np.log(10)/dOmega / Dpc**3 * dist_DDM
rhoH = Chalo * 5/np.log(10)/dOmega / Dpc**3 * dist_HDM

#Scale fractional error 
dxD_distN = rhoD * err_DDM / dist_DDM
dxH_distN = rhoH * err_HDM / dist_HDM
```

```python
#Define helper function for fitting exp. disk from Ivezic
def ExpDiskDensity(distance, A, H): # ln(rho) = A - H*distance
    return A - H*distance
```

```python
#Select region of log(rho_disk) vs distance to fit an exponential profile using scipy curve_fit()
xFitExpDisk = Dpc[np.logical_and((Dpc > 100),(Dpc < 500))]
yFitExpDisk = np.log(rhoD[np.logical_and((Dpc > 100),(Dpc < 500))])
A, H = curve_fit(ExpDiskDensity, xFitExpDisk, yFitExpDisk)[0]
rhoD0 = np.exp(A) 
print 'rhoD0 = {0}'.format(rhoD0) 
print 'scale height H = {0} pc'.format(1.0/H)

#Scale halo values by median value since function is roughly constant except for outliers near 0
rhoH0 = np.median(rhoH[(Dpc > 500) & (Dpc < 2000)])
print 'rhoH0 = {0}'.format(rhoH0)
 
#Renormalize luminosity function to units of number of stars per pc^3 AND per mag at D = 0 pc
yD_distN = dist_DMr * rhoD0
yH_distN = dist_HMr * rhoH0 

#Update fractional errors
dyD_distN = yD_distN * err_DMr / dist_DMr
dyH_distN = yH_distN * err_HMr / dist_HMr
```

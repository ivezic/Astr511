```python
from scipy.optimize import curve_fit

# Calculate the histogram ofthe distance modulus difference 
hist, bin_edges = np.histogram(diff, bins=50, density=True)
bin_centres = (bin_edges[:-1] + bin_edges[1:])/2

# Fit a gaussian to the distribution  from http://stackoverflow.com/questions/11507028/fit-a-gaussian-function  :
# Define model function to be used to fit to the data above:

def gauss(x, *p):
    A, mu, sigma = p
    return A*np.exp(-(x-mu)**2/(2.*sigma**2))

# p0 is the initial guess for the fitting coefficients (A, mu and sigma above)
p0 = [1., 0., 1.]

coeff, var_matrix = curve_fit(gauss, bin_centres, hist, p0=p0)

# Get the fitted curve
hist_fit = gauss(bin_centres, *coeff)


# Plot the result 
fig1, ax1 = plt.subplots(figsize=(12,10))
ax1.set_xlabel(r'$\Delta \mu $')
ax1.set_ylabel('Number of objects in the subsample')
ax1.set_title('Distance modulus difference')

ax1.plot(bin_centres, hist, ls = 'steps', label='Test data')
ax1.plot(bin_centres, hist_fit, label='Fit 1')


# Calculate statistics...
print 'Without clipping, the statistics are:'
print 'Fitted mean = ', coeff[1], 'Calculated mean', np.mean(diff)
print 'Fitted standard deviation = ', coeff[2], ' Calculated std dev ', np.sqrt(np.var(diff))
print 'Calculated median ', np.median(diff)
print 'Calculated rms ', np.sqrt(np.mean(np.square(diff)))
# Apply cuts 

# Calculate the sigma : standard deviation

sigma = np.sqrt(np.var(diff))
max_clip = coeff[1] + 3 * sigma
min_clip = coeff[1] - 3* sigma

clipping_mask = ( diff < max_clip ) * (diff > min_clip)


#  Fit a new gaussian to clipped data 

# Calculate the histogram ofthe distance modulus difference 
hist, bin_edges = np.histogram(diff[clipping_mask], bins=50, density=True)
bin_centres = (bin_edges[:-1] + bin_edges[1:])/2

# p0 is the initial guess for the fitting coefficients (A, mu and sigma above)
p0 = [1., 0., 1.]

coeff, var_matrix = curve_fit(gauss, bin_centres, hist, p0=p0)

# Get the fitted curve
hist_fit = gauss(bin_centres, *coeff)

# Plot the result 

ax1.plot(bin_centres, hist_fit, label='Fit 2')
ax1.legend()

# Report the results:   when calculating median and rms scatter, include only those +/- 3 * sigma  
print '\n With clipping, the statistics are : '
print 'Fitted mean',coeff[1], 'Calculated Mean', np.mean(diff[clipping_mask])
print 'Fitted std dev', coeff[2], ' Calculated std dev, i.e. new sigma', np.sqrt(np.var(diff[clipping_mask]))
print 'Calculated Median', np.median(diff[clipping_mask])
print 'Calculated rms ', np.sqrt(np.mean(np.square(diff[clipping_mask]))) #http://stackoverflow.com/questions/5613244/root-mean-square-in-numpy-and-complications-of-matrix-and-arrays-of-numpy
```

## B) "Gold parallax sample" : photometric parallax relation

### $gObs-rObs$ vs. $Mobs$ photometric parallax relation

Equations used : 
$D_{obs, kpc} = 1 / piObs $ , and thus $DM_{obs} = 5 log_{10}(1000 D_{obs, kpc}) - 5$ .  Then  $M_{obs} = rObs - DM_{obs}$.


To get error on  $M_{obs}$ I want to combine error of $rObs$  and $DM_{obs}$ in quadrature. I have explicit $rErr$, and $DM_{obs} = 10- 5 log(piObs) $, so that it has a form  $x = a log(u) + b$, and $\delta x   = a \delta u /  (u * ln(10)) $, i.e.  $\sigma _{x} = a * \sigma_{u} / ( 2.303 * u) $. 

Thus $\sigma_{DM} = 5 * \sigma_{piObs} / (piObs * 2.303) $, and 

$$\sigma_{M,obs} = \sqrt { \sigma_{rObs} ^{2} + \sigma_{DM}^{2} } $$

```python
from scipy.stats import binned_statistic
import numpy.polynomial.polynomial as poly
import matplotlib.gridspec as gridspec

# Define the sample 

T_mask =  truth_data['T'] == 1.0                               # i.e. hydrogen white WDs
mask_SN =  ( obs_data['piObs'] / obs_data['piErr'])  > 10.0    # high S/N ratio
ttl_mask = T_mask * mask_SN  

# Compute distance and distance modulus from the parallax measurement 
piObs = obs_data['piObs'][ttl_mask]
D_kpc_obs = 1.0 / piObs
DM_obs = 5 * np.log10(D_kpc_obs * 1000.0 ) - 5.0  # needed D in pc instead of kpc 

# Calculate r-band absolute magnitude 
Mobs = obs_data['rObs'][ttl_mask] - DM_obs
DM_obs_err = 5 * obs_data['piErr'][ttl_mask] / (obs_data['piObs'][ttl_mask] * 2.303)
Merr = np.sqrt(np.square(obs_data['rErr'][ttl_mask]) + np.square(DM_obs_err))

# Plot  absolute r-band magnitude based on DM derived from parallax,  and g-r Obs
gObs = obs_data['gObs'][ttl_mask] 
rObs = obs_data['rObs'][ttl_mask]
g_r = gObs - rObs

### PLOTTING #### 

# Define the entire plotting arrangement
fig1 = plt.figure(figsize=(12,10))
gs = gridspec.GridSpec(nrows=3, ncols=1, height_ratios=[4, 1,1]) 

# 1st plot  : g-r  vs. Mobs 
ax1 = plt.subplot(gs[0])
ax1.set_ylabel('$M_{r}$', fontsize=20)
ax1.set_title( ' g-r color vs. absolute r-band magnitude  ')
ax1.scatter( g_r, Mobs, s=0.4, alpha=0.4 )
plt.gca().invert_yaxis()
ax1.grid()
ax1.tick_params( axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='off')

#### BINNING  ##### 

# Bin g-r and compute median Mr in those bins
bin_width = 0.1 
nbins = int(max(Mr) - min(Mr)) / bin_width
bin_Mobs_means = binned_statistic(g_r, Mobs, statistic = 'median', bins=nbins)[0]
bin_Merr_means = binned_statistic(g_r, Merr, statistic = 'median', bins=nbins)[0]
# Pull out g-r at the bin edges where binned_statistic calculates mean ... 
bin_g_r_means = binned_statistic(g_r, Mobs, statistic = 'median', bins=nbins)[1][:-1]

#  1st plot  : add the mean points :
ax1.scatter(bin_g_r_means, bin_Mobs_means, s=23.5, alpha=1.0, color='yellow', edgecolors='black')

#### FITTING  ##### 

x = bin_g_r_means 
y = bin_Mobs_means
y_err = bin_Merr_means

sort= np.argsort(x)
x = x[sort]
y = y[sort]
y_err = y_err[sort]

# Problem with NaNs - I have no idea why binned_statistic would reuturn nans in some cases. 
# There are definitely no NaNs  in  Mobs,  nor in g_r ....  Remove them here manually 
indices = np.where((np.isnan(x) + np.isnan(y)) == False)[0]
x = x[indices]
y = y[indices]
y_err = y_err[indices]

####
# Fit a low order polynomial : try 2nd 
####

coefs = poly.polyfit(x, y, 2)
ffit = poly.polyval(x, coefs)
print 'Coefficients for 2nd order polynomial a + b*x + c*x^2  are a,b,c', coefs

# Plot the fit ... 
ax1.plot(x, ffit, 'r', lw=1)

# Plot residuals to the first fit  
ax2 = plt.subplot(gs[1])
ax2.set_ylabel('Res 2nd', fontsize=15)  
ax2.scatter(x,ffit-y, s=21.5, alpha=0.7, color='red') 
ax2.axhline(y=0, color='black', lw=2)
ax2.set_yticks([-0.2,-0.1, 0, 0.1, 0.2])
ax2.set_yticklabels(['-0.2','', '0.0', '', '0.2'])
ax2.tick_params( axis='x', which='both',  bottom='off', top='off', labelbottom='off')  #http://worksofscience.net/matplotlib/tick_labels
ax2.set_ylim(top=0.3, bottom=-0.3)
ax2.grid(axis='x')

####
# Fit a 3rd order polynomial
####

coefs = poly.polyfit(x, y, 3)
ffit = poly.polyval(x, coefs)
print 'Coefficients for 3rd order polynomial are ', coefs
# Plot the fit...
ax1.plot(x,ffit, 'b', lw=1)

# Plot residuals ... 
ax3 = plt.subplot(gs[2])
ax3.set_xlabel('$g-r$', fontsize=20)
ax3.set_ylabel('Res 3rd', fontsize=15)
ax3.set_yticks([-0.2,-0.1, 0, 0.1, 0.2])
ax3.set_yticklabels(['-0.2','', '0.0', '', '0.2'])
ax3.axhline(y=0, color='black', lw=2)
ax3.scatter(x, ffit-y, s=21.5, alpha=1.0, color='b')
ax3.set_ylim(top=0.3, bottom=-0.3)
ax3.grid(axis='x')

fig1.subplots_adjust(hspace=0)

####
# Fit a several orders ...  http://stackoverflow.com/questions/18767523/fitting-data-with-numpy
####
n_store= []
rms_store = []
chi2_store=[]
chi2red_store=[]
BIC_store=[]
AIC_store=[]

for n in range(1,13):
    coefs = poly.polyfit(x, y, n)
    ffit = poly.polyval(x, coefs)
    #ax1.plot(x,ffit, 'g', lw=2)
    rms = np.sqrt(np.mean(np.square(y-ffit)))
    chi2 = np.sum(((ffit-y )/ y_err )**2.0) 
    chi2red = chi2 / (len(y) + float(n)-1.0)
    AIC = chi2 + 2.0 * float(n)
    BIC = chi2 + float(n) * np.log(len(y))
    #print 'chi2 of fit with n=',n, ' is ', chi2
    #print 'reduced chi2 of fit with n=',n, ' is ', chi2red
    #print 'AIC of fit with n=',n, ' is ', AIC
    #print 'BIC of fit with n=',n, ' is ', BIC
    #print 'RMS of  fit with n=',n, ' is ', rms
    
    
    n_store.append(n)
    chi2_store.append(chi2)
    chi2red_store.append(chi2red)
    AIC_store.append(AIC)
    BIC_store.append(BIC)
    rms_store.append(rms)

    
# Plot the fitted rms...

fig2 = plt.figure(figsize=(12,5))
ay1 = fig2.add_subplot(111)
ay1.plot(n_store,rms_store, '-o')
ay1.set_title('RMS of fitted polynomials')
ay1.set_ylabel('RMS')
ay1.set_xlabel('Order of Polynomial')
ay1.set_xlim(left=0.5, right=12.5)

# plot the BIC, AIC, chi2...

fig3 = plt.figure(figsize=(12,5))
ay2 = fig3.add_subplot(111)
ay2.plot(n_store, chi2_store, '-o', label = r'$\chi^{2}$')
ay2.plot(n_store, AIC_store, '-o', label = 'AIC')
ay2.plot(n_store, BIC_store, '-o', label = 'BIC')
ay2.semilogy()
ay2.legend()
ay2.set_title(r'Goodness-of-fit for polynomial order in  $M_{r}$ vs $(g-r)$ relation')
ay2.set_ylabel(r'$(\chi^2, BIC, AIC)$')
ay2.set_xlabel('Order of Polynomial')
ay2.set_xlim(left=0.5, right=12.5)
```

Running it for a range of models shows that going from n=1 to n=2 we have a biggest improvement in $\chi ^2$ value, further adding terms makes only very small improvement. $\chi ^2$ will always decrease, i.e. improve the fit, as we add more terms, i.e. provide more degrees of freedom, but this also makes the information much less useful as overparametrised models become degenerate. This means that many different parametrisations of the same overparametrised mode could fit exactly the same data. The way to tell when to stop adding further terms is to quantify how adding further terms improves the goodness of fit.

Another way of quantifying the goodness of fit are  Bayesian Information Criterion (BIC) and Akaike Information Criterion (AIC). They are defined as :

BIC = $\chi ^2$ + Nterm $\cdot$ ln(Ndata)

AIC = $\chi ^2$ + 2 Nterm 

where  $\chi ^2$ = $\sum_i \chi_{i} ^2 $ , with $\chi_{i} =  (y_{model,i}  - y_{measured, i})  \,/\,   (y_{measured \, err, i})$

Thus from the plots  of BIC, AIC and $\chi^{2}$ above,  I can say that the most optimal seems to be $2^{nd}$ (from BIC) or $3^{rd}$ (from AIC) order. Thus the equation for photometric parallax relation is either :

$$ Mobs = 12.66741548  + 4.15424057 *(g-r)  -0.68200858 * (g-r)^2 $$

or 

$$ Mobs = 12.70976417 +  4.07596103 *(g-r)  -1.15672721 *(g-r)^{2} +    0.6050706 * (g-r)^{3}$$

```python
# Calculate Signal-to-Noise Ratio from proper motion components and proper motion error;  require SNR > 3

SNR = np.sqrt(np.square(obs_data['muRAObs'])  + np.square(obs_data['muDecObs'])) /  obs_data['muErr']
SN_mask = SNR > 3

# Combine masks
mask_add = SN_mask * pm_mask
```

I calculate $Mr$ from $gObs-rObs$ using the fit in part B). I use the 2nd order polynomial, adopting 

$$ Mr = 12.66741548  + 4.15424057 *(g-r)  -0.68200858 * (g-r)^2 $$.

Given $Mr$, I get the distance modulus:  $DM_{obs} = rObs - Mr$, and thus 

$$ D_{pc} = 10 ^ {\left( \left( DM + 5 \right)\, / \,5 \, \right)}$$

```python
# Calculate Mr and thus Distance[pc]
g = obs_data['gObs'][mask_add]
r = obs_data['rObs'][mask_add]
g_r = g - r

Mr = 12.66741548  + 4.15424057 *g_r  -0.68200858 * (g_r**2.0)  # already for masked objects 
DM = r - Mr
D_pc = np.power(10, (DM+5.0) / 5.0)


```

For the tangential velocity, $v_{t}$, I use a formula :
    
$$ v_{t} = 4.74 * \mu * d$$

where $\mu$ is the proper motion in $[arcsec / year]$ (combined from  $\mu_{RA}$ and $\mu_{dec}$ added in quadrature), and $d$ is the distance in $parsecs$.  Given that the proper motion is given in $[miliarcsec / year]$, I convert that to the desired unit before using the formula [adopted from Prof. Pogge notes](http://www.astronomy.ohio-state.edu/~pogge/Ast162/Unit1/motions.html)

```python
# Compute the tangential velocity 
muRA = obs_data['muRAObs'][mask_add]
muDec = obs_data['muDecObs'][mask_add]

mu = 0.001 * np.sqrt(np.square(muRA) + np.square(muDec))  # factor converts miliarcsec to arcsec 
vt = 4.74 * mu * D_pc
```

  $$ n_{c}(DM_{0}) = \int_{DM_{min}}^{DM_{0}} \frac{dn_{c}}{d(dM)} d(dM) $$
  $$ \Phi_{c}(M_{0}) = \int_{M_{min}}^{M_{0}} \frac{d\Phi_{c}}{dM} dM$$

So we see that the disk follows the exponential profile, but the halo follows a profile that isn't that steep. The scale height  that I calculate below , of 282 pc, is similar to the observed one on the left panel. 

For the plot above,  I used   $\rho (D) = x\_distribution * ( C_{sky} / 4 \pi D^{2} ) $

Now , to plot the $M$ distribution at $D=D_{0}$,  call it $\psi(M|D=D_{0})$, I use the fact that

$$ \psi(M|D=D_{N} )= \int \phi (M,D) dD =  \int \frac{C_{sky}}{4\pi D^{2}} \frac{d\phi_{c}}{dM} \frac{dn_{c}}{dD} dD = C_{sky} \frac{d \phi_{c}}{dM} = C_{sky} \cdot y\_distribution $$
and I extrapolate to $D_{0}$ : 
$$ \psi(M|D=D_{0} ) = \psi(M|D=D_{N} ) \cdot \frac{\rho(D_{0})}{\rho(D_{N})} = C_{sky} \cdot y\_dist \cdot \frac{x\_dist(D_{0})}{x\_dist(D_{N})}$$ 
with the last equality being valid because
$$ \rho(D_{i}) = x\_dist (D_{i}) \cdot \frac{D_{sky}}{4 \pi D^{2}}  $$ 

```python
# Calculating the luminosity function  

from scipy.optimize import curve_fit
# Calculating the luminosity function  at D=D0

xFitDisk = Dpc[(Dpc > 100) & (Dpc < 500)]
yFitDisk = np.log(rhoD[(Dpc > 100) & (Dpc < 500)])

def exp_Disk_rho(d, A, H): # ln(rho) = A - H*d
    return A - H*d

A, H = curve_fit(exp_Disk_rho, xFitDisk, yFitDisk)[0]
rhoD0 = np.exp(A) 
print 'Density_0 for Disk', rhoD0, ' scale height H (pc) =', 1/H

# For halo I take the median value for 500 < D/pc < 2000
rhoH0 = np.median(rhoH[(Dpc > 500) & (Dpc < 2000)])
print 'Density_0 for Halo', rhoH0

# Renormalize both halo and disk distributions
yD_distN = yD_dist * rhoD0
yH_distN = yH_dist * rhoH0 

# keep fractional errors remain 
dyD_distN = yD_distN * dyD_dist / yD_dist
dyH_distN = yH_distN * dyH_dist / yH_dist
```

Instead of $\psi$ I plot  $\phi$, which is (according to notes) the volume density: 
    
   $$ \phi (M,D) = \frac {d^{2}N}{dM dV} = \frac {d^{2}N}{dM dD} \frac{dD}{dV}, $$ and since $dV = 4 \pi D^{2} dD$, I get 
    
   $$ \phi (M,D) = \frac{C_{sky}}{4 \pi D^{2}} \cdot \frac{d \phi_{c}}{dM} \cdot \frac{d n_{c}}{dD} = \frac{C_{sky}}{4 \pi D^{2}} \cdot y\_dist \cdot x\_dist $$
